# The Forex Market 

The forex market is the largest financial market in the world, with a daily trading volume of trillions worth of foreign currencies. In this market, participants can buy, sell, exchange, and speculate on currencies. The simplest transaction is the exchange of two currencies at the current exchange rate. Market makers, such as banks, brokers, and other financial institutions, act as intermediaries, providing liquidity to the market. This leads to the **bid-ask spread**, which is the difference between the buying and selling prices of a currency pair.

## Needs in the Market: Derivatives

Market participants often need to do more than just exchange currencies. They must manage risks over time when dealing with **large transactions** or **long-term exposure** to currency fluctuations. For instance, consider a multinational company in Europe (let's call it MultiTech) that invoices U.S. customers in USD while incurring production costs in EUR. If MultiTech expects a $100 million payment in 3 months at the current exchange rate of 1.1000 EUR/USD, it anticipates receiving **€90.91 M**. However, if the USD weakens to **1.2500 EUR/USD**, MultiTech would receive only **€80 M**, leading to a loss of around **€10.91 M** purely due to currency fluctuations. Such a loss could erode investor confidence, drive up product prices, reduce demand, and potentially cause stock price declines. 

To manage these risks, **derivative instruments** have emerged. A derivative is a financial contract whose value is derived from an underlying asset, such as a currency pair. The most common derivatives in the Forex market are **futures**, **forwards**, and **options**.

### Futures, Options, and Beyond

A **futures contract** is an agreement to buy or sell a currency at a future date at a price agreed upon today. Futures are standardized and traded on exchanges. A **forward contract** is similar to a futures contract but is traded over-the-counter (OTC), allowing for more customization. 

An **option** gives the holder the **right**, but not the obligation, to buy (call) or sell (put) a currency at a predetermined strike price by a certain date. Managing and pricing these derivatives requires sophisticated mathematical models, as market makers need to quote accurate prices while taking the opposite side of the trades.

### European Options: Calls and Puts

The **Call option** gives the buyer the right (but not the obligation) to **buy** the underlying asset from the seller at a maturity time $T$ and a strike price $K$. In exchange for this right, the buyer pays the seller a premium, denoted by $C(t),\ t < T$. The **Put option** gives the buyer the right (but not the obligation) to **sell** the underlying asset to the seller at maturity time $T$ at a strike price $K$. The buyer pays the seller a premium, $P(t),\ t < T$, for this right. The difference between **European** and **American** options is that **European options** can only be exercised at maturity, while **American options** can be exercised at any time before maturity.

Let $S(t)$ be the price of the underlying asset at time $t$. The payoff of a call option at maturity is:

$$C(T) = \max(S(T) - K, 0)$$

The payoff of a put option at maturity is:

$$P(T) = \max(K - S(T), 0)$$

In a call option:
- If $S(T) > K$, the buyer exercises the option and buys the asset at $K$, making a profit of $S(T) - K$.
- If $S(T) < K$, the buyer will not exercise the option and loses only the premium $C(t)$.

In a put option:
- If $S(T) < K$, the buyer exercises the option and sells the asset at $K$, making a profit of $K - S(T)$.
- If $S(T) > K$, the buyer will not exercise the option and loses only the premium $P(t)$.

**How MultiTech could take leverage from a sell option** 

If MultiTech purchases a put option with a strike price of 1.1000 EUR/USD, and the USD weakens to 1.2500 EUR/USD, the company can exercise the option to sell at a higher price than the current market price, mitigating the loss from the currency devaluation. 
The following code makes a plot of the payoff of a call and put option:


```python
from matplotlib import pyplot as plt
import numpy as np

K = 1.10 # strike price
S = np.linspace(0.9, 1.3, 100) # underlying asset price

# call option
call_payoff = np.maximum(S - K, 0)

# put option
put_payoff = np.maximum(K - S, 0)

# plot the payoff of the call and put option
plt.figure(figsize=(10, 5))
plt.plot(S, call_payoff, label='Call Option Payoff')
plt.plot(S, put_payoff, label='Put Option Payoff')
plt.xlabel('Underlying Asset Price at Maturity')
plt.ylabel('Profit')
plt.title('Call and Put Option Payoff Diagram')
plt.legend()
plt.grid(True)
plt.show()
```

### Arbitrage

Arbitrage can be seen as a trading opportunity with a guaranteed profit without risk. In a arbitrage opportunity, the profit is based on the difference in prices of the same asset in different markets. The idea is to buy and sell the same asset in different markets at the same time. For example, if we have the following spot prices for three different pairs:

EUR/USD = 1.1000
USD/GBP = 0.8500
EUR/GBP = 0.9000

If we have €10, we can get 11 (10 * 1.1000) and then convert it to GBP, which gives us 11 * 0.8500 = £9.35. This £9.35 are more than enough to cover the €10 we spent to get it since £9.35 $\approx$ €10.39. In practice, the arbitrage opportunity will not be this easy, since the lack of liquidity, the quick movement of the market, and the bid-ask spread will make it difficult to earn money.



### Forwards and Futures

A **forward contract** is an agreement between two parties to buy or sell an asset at a future date at a price agreed upon today. The forward price is the price at which the asset will be bought or sold at the future date. The forward price is determined in such a way that the price for taking the opposite side of the trade is zero. The reason the forward contract value is zero at the start is to ensure that the contract is a fair and balanced agreement. The forward contract are traded over the counter (OTC), which means that there is not a third party involved in the contract. In forward contracts, there is a bigger credit risk because of the potential that either party could default on their obligations
 
A **future contract** is a standardized forward contract that is traded on an exchange. Futures are market to market contracts settled at the daily basis, since the value of the contract is adjusted daily to reflect the current market conditions. This daily settlement process ensures that any profits or losses from changes in the price of the futures contract are realized and accounted for on a daily basis. The credit risk is lower than in the forward market because of the exchange's role in the contract.
