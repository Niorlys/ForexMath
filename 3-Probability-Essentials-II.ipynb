# Random Variables 
 
When performing experiments, often we are interested in a quantitative characterization of the outcomes instead of the outcomes themselves. For example, when we toss two dice, we might be interested in the sum of the numbers that come up. In this case, we can build a function $X$ where $X(\omega) = X((i,j)) = i+j$. Then, for instance, the event $A = \{(1,2), (2,1)\}$ can be stated as $\{ \omega: X(\omega) = 3\} = X^{-1}(\{3\})$. Another example, when someone is betting on a coin toss, earning the amount bet if the coin lands heads and losing double the amount bet if the coin lands tails. In this case, for an amount $n$, the net return can be stated as $X_n = n I_H - 2n I_T = n(3I_H -2)$, where $I_H$ is the indicator function of the event where the coin lands heads and $I_T$ is the indicator function of the event where the coin lands tails, and $I_T = 1 - I_H$. Informally, we say that a **random variable** is any real-valued function on the sample space $\Omega$. For assigning probabilities to the events mapped by $X$, we must allow $X$ to satisfy the "natural" condition specified by the following definition.  

>**Definition 2.1 (Measurable function)** Let $(\Omega, \mathcal{F})$ be a measurable space. A function $f: \Omega \to \mathbb{R^p}$ is  $\langle \mathcal{F}, \mathcal{B}(\mathbb{R^p}) \rangle$ measurable, if for every Borel set $B \subset \mathbb{R^p}$, we have
$$f^{-1}(B) = \{ \omega \in \Omega: f(\omega) \in B \} \in \mathcal{F}.$$  
 
Often, for simplicity of notation, we will write $\{ \omega \in \Omega: f(\omega) \in B \}$ as $\{f \in B\}$.
 
>**Definition 2.2 (Random Variable)**  Let $(\Omega, \mathcal{F})$ be a measurable space. A function $X: \Omega \to \mathbb{R^p}$ is called a random variable if it is $\langle \mathcal{F}, \mathcal{B}(\mathbb{R^p}) \rangle$ measurable. This is; for every Borel set $B \in \mathcal{B}(\mathbb{R^p})$, we have 
$$X^{-1}(B) = \{ \omega \in \Omega: X(\omega) \in B \} \in \mathcal{F}$$

  
For simplicity, we take $p=1$ in the definition above, and when there is no ambiguity, we say only that $X$ is a random variable, assuming that preimages are taken with respect to the Borel $\sigma$-algebra in $\mathbb{R^p}$. We call the condition "natural" because, as we saw earlier, events defined through a random variable are described by the preimage of Borel sets. Since we assign a probability measure to these events, it is natural to require that the preimages of Borel sets be measurable. An inmediate example of randm variable is the indicator function of an event $A \subset \Omega$, given by $I_A(\omega) = 1$ if $\omega \in A$ and $I_A(\omega) = 0$ otherwise. If $B$ is any Borel set, then

$$
I_A^{-1}(B) = 
\begin{cases}
A, & \text{if } 1 \in B \text{ and } 0 \notin B,\\
A^c, & \text{if } 1 \notin B \text{ and } 0 \in B,\\
\varnothing, & \text{if } 1 \notin B \text{ and } 0 \notin B,\\
\Omega, & \text{if } 1 \in B \text{ and } 0 \in B.
\end{cases}
$$ 
 
Since all the preimages of $I_A$ are in $\mathcal{F}$, then $I_A$ is a random variable. However, verifying that every possible preimage of $X$ belongs to $\mathcal{F}$  as we did for $I_A$ is impractical. Fortunately, we can check whether $X$ is a random variable by ensuring that this condition holds for a class $\mathcal{C}$ of sets that generate the Borel $\sigma$-algebra on $\mathbb{R}$.

 
>**Theorem 2.1** 
> 
>Let $X: \Omega \to \mathbb{R}$ be a function, and let $\mathcal{C}$ a class of subsets of $\mathbb{R}$ that generates the Borel $\sigma$-algebra. Then, $X$ is a random variable if and only if for every $C \in \mathcal{C}$, we have
> $$X^{-1}(C) = \{ \omega \in \Omega: X(\omega) \in C \} \in \mathcal{F}.$$  

where $\mathcal{F}$ is the $\sigma$-algebra of the sample space $\Omega$.
  
**Proof:**  $\Rightarrow$ Let's define $\mathcal{G}=\{A \in \mathcal{B}(\mathbb{R}): X^{-1}(A) \in \mathcal{F}\}$. It is easy to prove that $\mathcal{G}$ is a $\sigma$-algebra. Since by assumption, $\mathcal{C} \subset \mathcal{G}$, we have that $\mathcal{B}(\mathbb{R}) = \sigma(\mathcal{C}) \subset \mathcal{G}\subset \mathcal{B}(\mathbb{R})$. Therefore, $\mathcal{G}=\mathcal{B}(\mathbb{R})$, and $X$ is a random variable. The other direction is trivial. 

$\blacksquare$ 
 
**Example 2.1** Let $\{X_i\}_{i=1}^n$ a finite collection of random variables respect to $\mathcal{F}$, with $X_i:\Omega\rightarrow\mathbb{R}^n$, then $X=(X_1,X_2,...,X_n)$ is a random variable respect to $\langle \mathcal{F}, \mathcal{B}(\mathbb{R}^n) \rangle$. 
 
Since $\mathcal{B}(\mathbb{R}^n)$ is generated by the class $\mathcal{C}$ of open rectangles  in $\mathbb{R}^n$, then let $C = \prod_{i=1}^n (a_i, b_i)$ be an open rectangle. Then, $\forall C \in \mathcal{C}$, we have
$$X^{-1}(C) = \prod_{i=1}^n X_i^{-1}((a_i, b_i)) = \bigcap_{i=1}^n \{ \omega: X_i(\omega) \in (a_i, b_i)\} \in \mathcal{F}$$ 
 
since $\{ \omega: X_i(\omega) \in (a_i, b_i)\}$ is in $\mathcal{F}$ for all $i=1,2,...,n$ due to the fact that $X_i$ is a random variable. Therefore, $X$ is a random variable by Theorem 1.1.   $\blacksquare$

An useful case is when $\mathcal{C}$ is the class of intervals of the form $(-\infty, x]$. Theorerem 1.1 from the first article states that Borel's $\sigma$-algebra is generated by the class $\mathcal{C} = \{(-\infty, x]: x \in \mathbb{R}\}$. This leads to the following corollary. 

>**Corollary 2.2**
>
>Let $X: \Omega \to \mathbb{R}$ be a function. Then, $X$ is a random variable if and only if for every $x \in \mathbb{R}$, we have
$$X^{-1}((-\infty, x]) = \{ \omega \in \Omega: X(\omega) \leq x \} \in \mathcal{F}.$$  

The above corollary is particularly useful to confirm that a function is a random variable, by stating that is sufficient to check if the preimage of intervals of the form $(-\infty, x]$ is in $\mathcal{F}$ as we show in the following example. 
 
 
 **Example 2.2** Let $\{X_n\}_{n\in\mathbb{N}}$ a sequence of random variables respect to $\langle \mathcal{F}, \mathcal{B}(\mathbb{\bar{R}}) \rangle$. Then, $X = \sup_{n\in\mathbb{N}} X_n$ is a random variable in $\langle \mathcal{F}, \mathcal{B}(\mathbb{\bar{R}}) \rangle$.   
  
For every $x \in \mathbb{R}$, we have
  
$$X^{-1}((-\infty, x]) = \{ \omega: X(\omega) \leq x \} = \bigcap_{n\in\mathbb{N}}\{\omega: X_n(\omega)\leq x\}=\bigcap_{n\in\mathbb{N}} X_n^{-1}((-\infty, x]) \in \mathcal{F}$$ 
 
since $X_n$ are random variables. Therefore, $X$ is a random variable by Corollary 1.2. The same argument can be used for the infimum, taking $\inf_{n\in\mathbb{N}} X_n = - \sup_{n\in\mathbb{N}} (-X_n)$. $\blacksquare$ 
 


  
>**Proposition 2.3** Let $X:\Omega\rightarrow\mathbb{R^p}$ be a random variable and $g:\mathbb{R^p}\rightarrow\mathbb{R^q}$ measurable in $\langle \mathcal{B}(\mathbb{R^p}), \mathcal{B}(\mathbb{R^q}) \rangle$. Then, $Y=g(X)$ is a random variable respect to $\langle \mathcal{F}, \mathcal{B}(\mathbb{R^q}) \rangle$.  
 
**Proof:** Left as an exercise. $\blacksquare$  
 
In order to apply the above proposition safely, we need to know when a real function is measurable. A sufficient condition is given by the following proposition.
 
>**Proposition 2.4** Let $g:\mathbb{R^p}\rightarrow\mathbb{R^q}$ be continuous. Then, $g$ is measurable in $\langle \mathcal{B}(\mathbb{R^p}), \mathcal{B}(\mathbb{R^q}) \rangle$.
 
**Proof:** Left as an exercise. Hint: use the fact that the preimage of an open set under a continuous function is open. $\blacksquare$ 
 
 **Example 2.3** If we have $n$ random variables $\{X_k\}_{k=1}^n$ respect to $\langle \mathcal{F}, \mathcal{B}(\mathbb{R}) \rangle$, then $Y = X_1 + X_2 + ... + X_n$ is a random variable in $\langle \mathcal{F}, \mathcal{B}(\mathbb{R}) \rangle$. 
  
To see this, recall by **Exampple 1.3** that $X=(X_1, X_2, ..., X_n)$ is a random variable respect to $\langle \mathcal{F}, \mathcal{B}(\mathbb{R}^n) \rangle$. Also, the function $g(x_1, x_2, ..., x_n) = x_1 + x_2 + ... + x_n$ is continuous in $\mathbb{R}^n$ and therefore measurable, then by **Proposition 1.5**, $Y=g(X) = X_1 + X_2 + ... + X_n$ is a random variable in $\langle \mathcal{F}, \mathcal{B}(\mathbb{R}) \rangle$. $\blacksquare$  
 
Another important examples of random variables are presented in the following proposition. The reader should be able to prove the following proposition by himself. 

>**Proposition 2.5** Let $X_n:\Omega\rightarrow\mathbb{R}$ be a random variable $\forall n\in\mathbb{N}$, then the following functions are random variables: 
>
>i. $\sup_{n\in\mathbb{N}} X_n$ 
>
>ii. $\inf_{n\in\mathbb{N}} X_n$ 
>
>iii. $\limsup_{n\in\mathbb{N}} X_n$ 
>
>iv. $\liminf_{n\in\mathbb{N}} X_n$ 
>

**Proof:** Left as an exercise. $\blacksquare$


 ## Distribution of a random variable
 Recall that we use random variables because most of the time we are interested in a quantitative characterization of the outcomes instead of the outcomes themselves. A random variable allows us to abstract ourselves from the original sample space and simply work with the Borel sets. We know that $X^{-1}(B)$ is measurable for every $B \in \mathcal{B}(\mathbb{R})$, so we can define a new probability measure on $\mathcal{B}(\mathbb{R})$ as follows: 
  
$$
P_X(B) = P(X^{-1}(B)) = P(\{ \omega \in \Omega: X(\omega) \in B \}) = P(X\in B)
$$

 It is easy to check that $P_X$ is a probability measure on $\mathcal{B}(\mathbb{R})$.  

 >**Definition 2.3 (Distribution of a random variable)** Let $X$ be a random variable on $(\Omega, \mathcal{F}, P)$. The distribution of $X$ is the probability measure $P_X$ on $\mathcal{B}(\mathbb{R})$. 
  
Some examples of distributions are:  

i. $P(X=k) = \displaystyle\frac{1}{k(k+1)}$ for $k=1,2,3,...$. Note that $\sum_{k=1}^{\infty} P(X=k) = 1$ since it is a telescoping series starting at $k=1$. 

ii. Let $B$ be a Borel set, $P(X\in B) = \displaystyle\frac{2}{\pi} \int_{B\cap\mathbb{R}^+} \frac{1}{1+x^2} dx$. Take into account that $\displaystyle\int_{0}^{\infty} \frac{1}{1+x^2} dx = \frac{\pi}{2}$.

 So we can see that any random variable induces a probability measure on $\mathcal{B}(\mathbb{R})$. An important use of this concept through the cumulative distribution function (CDF) of $X$, defined as
 
 $$
 F_X(x) = P_X((-\infty, x]) = P(X\leq x), \quad x \in \mathbb{R}
 $$ 
  
**Example 2.4 (Cumulative effect of $F_X$)** Let $X$ be a random variable on $\Omega = \{1,2,3\}$ with $P(X=1) = 0.1$, $P(X=2) = 0.3$ and $P(X=3) = 0.6$. Then, some values of the CDF of $X$ are given by
 
 - $F_X(a) = 0$ for $a < 1$
 - $F_X(1) = 0.1$
 - $F_X(2) = 0.1 + 0.3 = 0.4$
 - $F_X(3) = 0.1 + 0.3 + 0.6 = 1$
 - $F_X(b) = 1$ for $b > 3$ 
  
  $\blacksquare$  
   
>**Proposition 2.6 (Properties of the CDF)** Let $F_X$ be the CDF of a random variable $X$. Then,
>
>i. $F_X$ is non-decreasing.
>
>ii. $F_X$ is right-continuous, i.e., $\lim_{x\to x_0^+} F_X(x) = F_X(x_0)$ for all $x_0 \in \mathbb{R}$.
>
>iii. $\lim_{x\to-\infty} F_X(x) = 0$ and $\lim_{x\to\infty} F_X(x) = 1$.

**Proof:** 
i. Let $x_1 < x_2$. Then, $(-\infty, x_1] \subset (-\infty, x_2]$. Therefore, $F_X(x_1) = P(X\leq x_1) \leq P(X\leq x_2) = F_X(x_2)$.

ii. Let $x_n \downarrow x_0$. Then, $(-\infty, x_n] \downarrow (-\infty, x_0]$. Therefore, by the continuity of $P$, we have $F_X(x_n) \downarrow F_X(x_0)$.

iii. Let $x_n \to -\infty$. Then, $(-\infty, x_n] \to (-\infty, -\infty] = \varnothing$. Therefore, $F_X(x_n) \to 0$. Let $x_n \to \infty$. Then, $(-\infty, x_n] \to (-\infty, \infty) = \mathbb{R}$. Therefore, $F_X(x_n) \to 1$.


$\blacksquare$

In measure theory, every function $F$ satisfying the properties in the proposition above is called a **distribution function**. A useful result valuable on integration is that given the CDF $F_X$ of a random variable $X$, we can recover the probability measure $P$ on $\mathcal{B}(\mathbb{R})$ as follows:
 $$
 P_X((a,b]) = F_X(b) - F_X(a+),\\
 P_X((a,\ \infty)) = 1 - F_X(a+),\\
 $$
  for all $a,b \in \mathbb{R}$ with $-\infty \le a < b < \infty$. Then the following holds:

  **Theorem 2.7** Let $F$ a distribution function on $\mathbb{R}$. Then, there exists a unique probability measure $P$ on $\langle \mathbb{R}, \mathcal{B}(\mathbb{R}) \rangle$ such that $P_F((a,b]) = F(b) - F(a)$ for all $a,b \in \mathbb{R}$ with $-\infty \le a < b < \infty$. The measure $P_F$ is called the Lebesgue-Stieltjes measure induced by $F$.
  
  **Proof:** The proof of this can be found in Chapter 2, Section 2.3 of "Probability-1" from Albert N. Shiryaev. The classical construction extends $P$ from the class $\mathcal{C}$ of half-open intervals to the Borel $\sigma$-algebra $\mathcal{B}(\mathbb{R})$ by means of the Caratheodory extension theorem. $\blacksquare$
  
Since CDFs are distributions, then $P_{F_X} = P_X$. **Caution:** in probability theory, the notation $dF_X$ in the context of integration does not represents a differential, but a notation for the Lebesgue-Stieltjes measure induced by $F_X$.
 
 By continuity of $P$ from below, the reader should be able to prove that $P(X<x) = F_X(x-)$ for all $x \in \mathbb{R}$. Since $\{X\le x\} = \{X<x\} \cup \{X=x\}$, we have $P(X\le x) = P(X<x) + P(X=x)$. Therefore, $P(X=x) = P(X\le x) - P(X<x) = F_X(x) - F_X(x-)$. This way, we see that $F_X$ has jumps at the points $x$ where $P(X=x) > 0$. From real analysis, a monotone function can have at most a countable number of discontinuities. Therefore, the set
 $\{x: P(X=x) > 0\}$ is countable. So, naturally we were led to the following definitions.

>**Definition 2.4 (Discrete random variable)** A random variable $X$ is called discrete if its distribution is concentrated on a countable set, i.e., if there exists a countable set $A \subset \mathbb{R}$ such that $P(X \in A) = 1$.

**Example 2.5** Let $X$ be a random variable denoting the number obtained when a fair die is rolled. Then, $X$ is a discrete random variable since it takes values in the set $A = \{1,2,3,4,5,6\}$,, and $P(X=x) = \frac{1}{6}$ for all $x \in A$. It is trivial that $P(X\in A) = 1$.
 
>**Definition 2.5 (Continuous random variable)** A random variable $X$ is called continuous if $P(X=x) = 0$ for all $x \in \mathbb{R}$. 

**Example 2.6** Let $X$ be the amount of time (in seconds) we have to wait for to register the emission of a radioactive particle from a source that emits particles within a time interval of 1 second. Then, $X$ is a continuous random variable in $(\mathbb R, \mathcal{B}(\mathbb{R}), P)$, with $P(X\in\ (0,\ x]) = x$. This distribution is called **uniform distribution** on $(0,1)$.
  
Next code shows plots of the CDF of a continuous and a discrete random variable example that we will see later.
  
 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, bernoulli

x_cont = np.linspace(-4, 4, 500)
x_disc = np.array([-1, 0, 1, 2])

# Continuous CDF: Normal distribution
mu, sigma = 0, 1
cdf_cont = norm.cdf(x_cont, mu, sigma)

# Discrete CDF: Bernoulli(p)
p = 0.7
cdf_disc = bernoulli.cdf(x_disc, p)


fig, axs = plt.subplots(1, 2, figsize=(12, 4))

axs[0].plot(x_cont, cdf_cont, color='green')
axs[0].set_title('Continuous CDF')
axs[0].set_xlabel('x')
axs[0].set_ylabel('F(x)')
axs[0].grid(True)

axs[1].step(x_disc, cdf_disc, where='post', color='green')
axs[1].set_title('Discrete CDF')
axs[1].set_xlabel('x')
axs[1].set_ylabel('F(x)')
axs[1].set_xticks(x_disc)
axs[1].grid(True)

plt.tight_layout()
plt.show()
```


### Independence of random variables

The concept of independence can be extended to random variables, with the idea that for tow random variables $X$ and $Y$, the probability of $X$ taking a value in a Borel set $A$ is independent of the probability of $Y$ taking a value in a Borel set $B$. Formally, we have the following definition.

> **Definition 3.1 (Independence of random variables)** Two random variables $X$ and $Y$ are said to be independent if for all $A,\ B \in \mathcal{B}(\mathbb{R})$, we have that 
>$$P(X\in A, Y\in B) = P(X\in A)P(Y\in B).$$

Since the Borel sigma-algebra is generated by the sets of the form $(-\infty, x]$, then the definition above is equivalent to say that for all $x,y \in \mathbb{R}$, we hvae the following proposition.

> **Proposition 3.1 (Independence of random variables)** Two random variables $X$ and $Y$ are independent if and only if for all $x,y \in \mathbb{R}$, we have that
>$$P(X\le x, Y\le y) = P(X\le x)P(Y\le y).$$
> In general, we say that a collection of random variables $\{X_i\}_{i\in I}$ is independent if for all Borel sets $A_i \in \mathcal{B}(\mathbb{R})$ for all $i\in I$, we have that
>$$P\left(\bigcap_{i\in I} (X_i \le x_i)\right) = \prod_{i\in I} P(X_i \le x_i).$$

Note that the last statement is the same as
$$
F_{(X_1,X_2,...,X_n)}(x_1,x_2,...,x_n) = F_{X_1}(x_1)F_{X_2}(x_2)...F_{X_n}(x_n).
$$

where $F_{(X_1,X_2,...,X_n)}$ is the joint CDF of the random variables $\{X_i\}_{i\in I}$. 
 
An usefult concept about random variables is the concept of **indepentent and identically distributed** random variables.

> **Definition 3.2 (Independent and identically distributed random variables)** A collection of random variables $\{X_i\}_{i\in I}$ is said to be independent and identically distributed (i.i.d.) if it is independent and all random variables have the same distribution.


## Integration of random variables: Expectation

From school we all have an intuitive understanding of the "average", formally known as the arithmetic mean. Our final score was computed as an average of our scores in the different subjects, that way future reviewer of our academic performance could have an idea about how well(or bad) we did. In the end, the average
$$
\bar{x} = \frac{1}{n} \sum_{i=1}^n x_i
$$
is a way to summarize the dataset into a single, representative number, aiming to capture the central tendency of that data. Now, suppose that there are values repeated in the dataset, let's say that the value $x_i$ occurs $n_i$ times in a dataset of size $m\ge n$. Then, the average can be rewritten as
$$
\bar{x} = \frac{1}{m} \sum_{i=1}^n n_i x_i = \sum_{i=1}^n \frac{n_i}{m} x_i
$$
where $\frac{n_i}{m}$ is the relative frequency of the value $x_i$ in the dataset. This way, the average can be seen as a weighted sum of the values in the dataset, where the weights are the relative frequencies of the values. In the long run, we can replace the relative frequencies by the probabilities of the values $x_i$ which can be interpreted as the values of a discrete random variable $X$. If we set $A_i=\{X=x_i\}$, then $X=\sum_{i=1}^n x_i \mathbb{1}_{A_i}$. Variables of this form are called **simple** as we define next, assuming that we have a probability space $(\Omega, \mathcal{F}, P)$. In what follows, we use the terms **Lebesgue integral** and **expectation** interchangeably.

> **Definition 2.5 (Simple random variable)** A random variable $X$ is called simple if it can be written as $\mathbb{S}=\sum_{i=1}^{n}x_i\mathbf1_{A_i}$, where $x_i \in \mathbb{R}$ and $A_i \in \mathcal{F}$ for all $i=1,2,...,n$.

Note that $\mathbb{S}$ can have different representations. For instance, $\mathbb{S}=I_{(0,1]} + 2I_{(1/2,2]} = I_{(0,1/2]} + 3I_{(1/2,1]} + 2I_{(1,2]}$.

> **Definition 2.6 (Expectation of a simple random variable)** Let $\mathbb{S}=\sum_{i=1}^{n}s_i\mathbf1_{A_i}$ be a simple r.v. on $(\Omega,\mathcal F,P)$. Its **expected value** (Lebesgue integral) is:
> $$
> \mathbb E[\mathbb{S}] \;=\;
> \int_{\Omega}\mathbb{S}\,\mathrm dP
> := \sum_{i=1}^{n}s_i\,P(A_i)
> $$

For example, the expected value of an indicator function of an event is the probability of the event, this is, if $A\in\mathcal F$, then $\mathbb E[1_A] = P(A)$. The definition of the integral mirrors the weighted average idea from the introduction: each value $x_i$ is weighted by the probability of the event where it occurs. From now on, we adopt the following conventions: 

 - Let $S(x)$ be a logical proposition. Then, we say that $S(x)$ is true almost surely (a.s.) if $P(S(x)\text{ is false}) = 0$.
 
The reader may verified that the integral of a simple random variable is well-defined, this is, it is independent of the representation of $\mathbb{S}$. Also, the properties of this integral are simple to verify, and we list them in the following proposition.

> **Proposition 2.7 (Properties of the integral)** Let $\mathbb{S}$ and $\mathbb{T}$ be simple random variables. Then,
> 
> 1. (Linearity) For any positives constants $a$ and $b$, we have $\mathbb{E}[a\mathbb{S} + b\mathbb{T}] = a\mathbb{E}[\mathbb{S}] + b\mathbb{E}[\mathbb{T}]$.
> 2. (Monotonicity) If $\mathbb{S} \leq \mathbb{T}$ a.s., then $\mathbb{E}[\mathbb{S}] \leq \mathbb{E}[\mathbb{T}]$.
> 3. (Positivity) If $\mathbb{S} = \mathbb{T}$ a.s., then $\mathbb{E}[\mathbb{S}] = \mathbb{E}[\mathbb{T}]$.
> 

Now we have defined the integral of basic kind of random variables (r.v), we proceed to define the integral of more general r.v. by approximating it using simple r.v and then making a limiting process. Let's start with the case of non-negative r.v. $X$. We want to approximate $X$ using simple r.v. with error less than $1/2^n$. The approximation procedure is as follows: 
 
1. For a given $n\in\mathbb{N}$, we partition the range of $X$ into intervals of length $1/2^n$.
2. For each $y = X(\omega)$, if $y\ge n$, then we approximate $y$ by $n$. If $y<n$, then $y$ must be in one of the intervals of length $1/2^n$, this is; $k/2^n \le y < (k+1)/2^n$, for some integer $k=0,1,2,...,n2^n-1$. Then, we approximate $y$ by the infimum $k/2^n$. Note that if we multiply by $2^n$ in the last inequality, we get $k \le 2^n y < k+1$, which is equivalent to $k = \lfloor 2^n y \rfloor$. This leads to the following simple r.v. :
$$
\mathbb{S}_n(\omega) = \begin{cases}
n, & \text{if } X(\omega) \ge n, \\
\displaystyle\frac{\lfloor 2^n X(\omega) \rfloor}{2^n}, & \text{if } X(\omega) < n.
\end{cases}
$$

The following script plots the approximation of a non-negative random variable $X$ using simple r.v. with error less than $1/2^n$.


```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 2 * np.pi, 500)
X = np.sin(t) + 1          # nonnegative; 0  X  2

# Simple-function builder  S() = floor(min(X,n)2)/2
def simple_approximation(x: np.ndarray, n: int) -> np.ndarray:
    """Return the simple-function approximation S_n(x)."""
    capped = np.minimum(x, n)          # truncation at the level n
    return np.floor(capped * 2**n) / 2**n

# Plot X together with S for several n
n_values = [1, 3, 6]
plt.figure(figsize=(12, 7))
plt.plot(t, X, linewidth=2, label="True  $X(t)=\\sin t + 1$")

for n in n_values:
    S_n = simple_approximation(X, n)
    plt.step(t, S_n, where="post", linestyle="--", label=f"$S_{n}$  (n = {n})")

plt.title("Monotone simple-function approximations $S_n \\uparrow X$")
plt.xlabel("t")
plt.ylabel("Value")
plt.grid(alpha=0.4)
plt.yticks(np.arange(0, 2.25, 0.25))
xticks = np.linspace(0, 2 * np.pi, 9)
xtick_labels = ["0"] + [f"{x/np.pi:.1g}" for x in xticks[1:]]
plt.xticks(xticks, xtick_labels)
plt.legend()
plt.tight_layout()
plt.show()
```


To see that $\mathbb{S}_n$ is non decreasing, it is enough to note that refining the partion, the intervals in the form $[k/2^n, (k+1)/2^n)$ produce two new intervals: $[k/2^n, (2k+1)/2^{n+1})$ and $[(2k+1)/2^{n+1}, (k+1)/2^n)$. For $\mathbb{S}_n$, we have the following proposition:
 
 >**Proposition 2.8** Let $X$ be a non-negative r.v. Then, $\mathbb{S}_n(\omega) \uparrow X(\omega)$ for all $\omega\in\Omega$. 
  
**Proof:** Left as exercise. $\blacksquare$
 
 Now, we define the Lebesgue integral of a non-negative r.v. $X$ as the limit of the integrals of the simple r.v. $\mathbb{S}_n$ that approximate $X$.

> **Definition 2.9 (Expectation of a non-negative r.v.)**  For $X\ge 0$ measurable, let $(\mathbb{S}_k)_{k\ge1}$ a sequence of simple r.v.s such that  $0\le \mathbb{S}_1\le \mathbb{S}_2\le\cdots\uparrow X$ point-wise, then the Lebesgue integral of $X$ is given by
> $$
> \mathbb E[X]=\lim_{k\to\infty}\mathbb E[\mathbb{S}_k].
> $$

Equivalently, we can define the Lebesgue integral of $X$ as the supremum of the integrals of the simple r.v.s that are less than or equal to $X$.
$$
\mathbb E[X]=\sup\bigl\{
\mathbb E[S]:
0\le S\le X,\;S\ \text{simple}
\bigr\}.
$$


Note that the integral defined above is well-defined, since always we can find a sequence of simple r.v.s that approximate $X\ge 0$ from below. Also, we have the following consistency property:

> **Proposition 2.10** Let $X\ge 0$ be a non-negative r.v., and let $\mathbb{S}_k\uparrow X$, $\mathbb{T}_k\uparrow X$ be a sequence of simple r.v.s. Then,
> $$
> \mathbb E[X]=\lim_{k\to\infty}\mathbb E[\mathbb{S}_k]=\lim_{k\to\infty}\mathbb E[\mathbb{T}_k].
> $$

So, the integral of $X$ is independent of the sequence of simple r.v.s that we use to approximate it. The next ttheorem enumarates a set of useful properties of the integral of non-negative r.v.s., we prove some of them, the rest are left as exercises.

> **Theorem 2.11** Let $X\ge 0$ and $Y\ge 0$ be non-negative r.v. in a probability space $(\Omega, \mathcal{F}, P)$. Then,
> 
> 1. (Linearity) For any positives constants $a$ and $b$, we have $\mathbb{E}[aX + bY] = a\mathbb{E}[X] + b\mathbb{E}[Y]$. 
>
> 2. (Monotonicity) If $X \leq Y$ a.s., then $\mathbb{E}[X] \leq \mathbb{E}[Y]$. 
>
> 3. $X=0$ a.s if and only if $\mathbb{E}[X]=0$. (This equivalence can fail when $X$ is not non-negative.)
>
> 4. If $X = Y$ a.s., then $\mathbb{E}[X] = \mathbb{E}[Y]$.
>
> 5. (Monotone convergence theorem) If $0\le X_1\le X_2\le\cdots\uparrow X$, then $\mathbb{E}[X_n]\uparrow\mathbb{E}[X]$. 
>
> 6. Let $\{X_n\}_{n\in\mathbb{N}}$ be a sequence of non-negative r.v.s. Then, $\mathbb{E}[\sum_{n=1}^{\infty}X_n]=\sum_{n=1}^{\infty}\mathbb{E}[X_n]$. 
>
> 7. (Fatou's lemma)  Let $\{X_n\}_{n\in\mathbb{N}}$ be a sequence of non-negative r.v.s., then $\mathbb{E}[\liminf X_n]\le\liminf\mathbb{E}[X_n]$. 
> 
> 8. (Independence) Let $X_1,X_2,...,X_n$ be independent r.v.s. Then, $\mathbb{E}[X_1X_2...X_n]=\mathbb{E}[X_1]\mathbb{E}[X_2]...\mathbb{E}[X_n]$.

**Proof:**  
 
3: If $X=0$ a.s., let $A=\{X=0\}$. Then, $P(A^c)=0$ and $\mathbb{E}[X]=\mathbb{E}[0\cdot 1_{A}]+\mathbb{E}[X\cdot 1_{A^c}]=0$. Conversely, if $\mathbb{E}[X]=0$,and let $D_n=\{X>1/n\}$. Then, $D_n\uparrow A^c$ and $0 = \mathbb{E}[X]\ge\mathbb{E}[X\cdot 1_{D_n}]\ge \frac{1}{n}P(D_n)$. This implies that $P(D_n)=0$, so taking limit and using continuity from below, we get that $\lim_{n\to\infty}P(D_n) = P(\cup_{n=1}^{\infty}D_n) = P(X>0)=0$.  
 
6: If we define $Y_n=\sum_{i=1}^{n}X_i$, then $Y_n\uparrow \sum_{n=1}^{\infty}X_n$ and $Y_n\ge 0$ a.s. by the monotonicity of the integral. Then, by the monotone convergence theorem, we have that $\mathbb{E}[\sum_{n=1}^{\infty}X_n]=\mathbb{E}[\lim_{n\to\infty}Y_n]=\lim_{n\to\infty}\mathbb{E}[Y_n]=\sum_{n=1}^{\infty}\mathbb{E}[X_n]$. 
 
7: Let's define $Y_n(\omega)=\inf\{X_j(\omega):\ j\ge n\}$. The $Y_n\uparrow \liminf X_n$. In addition, since $X_n\ge Y_n$, from monoticity we have:

 Applying limit on the above inequality, and using monotone convergence theorem (m.c.t) in the right side we get the result.

What about the general case of a real-valued r.v. $X$? Any real-valued r.v. $X$ can be split into its positive and negative parts:
$$
X^+:=\max\{X,0\},\qquad
X^-:=\max\{-X,0\},\qquad
X = X^{+}-X^{-}.
$$
Since $X^+$ and $X^-$ are non-negative, we can use their Lebesgue integral to define the integral of $X$. 

> **Definition 2.12 (Integrable random variable)**  Let $X$ be a random variable, we say that it is **integrable** if $\mathbb E[X^{+}] + \mathbb E[X^{-}] <\infty$.  Then we set
> $$
> \mathbb E[X] := \mathbb E[X^{+}] - \mathbb E[X^{-}].
> $$

Note that in order to $\mathbb E[X]$ to exist (it could be $\infty$), it is necessary that $\min\{\mathbb E[X^{+}], \mathbb E[X^{-}] \} <\infty$, otherwise we get an indeterminate form. Also, we have that $|X| = X^+ + X^-$, which leads to the followin equivalent definition of integrability.
 
>**Definition 2.13** We say that $X$ is **integrable** if $\mathbb E[|X|]<\infty$.

> **Theorem 2.14** Let $X$ be an integrable r.v. Then, $|X|<\infty$ a.s. 
 
**Proof:** Left as exercise. Hint: Define $A_n=\{ X> n\}$ and use continuity from above (c.f.a) to show that $P(X=\infty)=0$. $\blacksquare$

>**Definition 2.15 (Integral over a set)** Let $A\in\mathcal{F}$ and $X$ be a real-valued r.v. We define the integral of $X$ over $A$ as
> $$
> \int_A X dP := \mathbb E[X\cdot 1_A].
> $$

>**Definition 2.16 ($\mathcal{L}^p(\Omega,\mathcal{F},P)$ spaces)** We define $\mathcal{L}^p(\Omega,\mathcal{F},P)$ as the set of all real-valued r.v.s $X$ such that $\mathbb E[|X|^p]<\infty$.

In order to be able to use benefits of tools derived from normed and Hilbert spaces, we need to comprise $\mathcal{L}^p(\Omega,\mathcal{F},P)$ useing the fact that $X = Y$ a.s. is an equivalence relation. Next we make a brief review of equivalence relations.
 
>**Definition 2.17 (Equivalence relation )** Let $\Omega$ be a non-empty set, a binary relation in $\Omega$ is called an **equivalence relation** denoted by $\sim$ if it satisfies the following properties:
>
> 1. (Reflexivity) $x\sim x$ for all $x\in\Omega$.
> 2. (Symmetry) For all $x,y\in\Omega$, if $x\sim y$, then $y\sim x$.
> 3. (Transitivity) For all $x,y,z\in\Omega$, if $x\sim y$ and $y\sim z$, then $x\sim z$. 

>**Definition 2.18 (Equivalence class)** Let $\Omega$ be a non-empty set, and let $\sim$ be an equivalence relation in $\Omega$. The **equivalence class** of $x\in\Omega$ is the set of all elements in $\Omega$ that are equivalent to $x$. That is,
> $$
> [x] = \{y\in\Omega: y\sim x\}.
> $$
> The set of all equivalence classes of $\sim$ is denoted by $\Omega/\sim$, called the **quotient set** of $\Omega$ by $\sim$.

The following proposition is a direct consequence of the definition of equivalence relation.

> **Proposition 2.19** The relation $X\sim Y$ if and only if $X = Y$ a.s. is an equivalence relation.

Now we use the above proposition to make a more useful definition of space of integrable random variables, by identifying the random variable with its equivalence class.

> **Definition 2.20** We define 
> $$
> \mathbb{L}^p(\Omega,\mathcal{F},P) := \Omega/\sim,
> $$
> where $\sim$ is the equivalence relation $X\sim Y$ if and only if $X = Y$ a.s.

In what follows, doing an abuse of notation, we will use $X$ to denote the equivalence class of $X$ in $\mathbb{L}^p(\Omega,\mathcal{F},P)$. Now we can define the norm in $\mathbb{L}^p(\Omega,\mathcal{F},P)$ as follows:

> **Definition 2.21** For $X\in\mathbb{L}^p(\Omega,\mathcal{F},P)$, and $0<p<\infty$:
> $$
> \|X\|_{p} := \mathbb E[|X|^p]^{\min(1/p,1)}.
> $$

Now we have a normed space, and we can define the distance between two random variables as follows:

> **Definition 2.22** For $X,Y\in\mathbb{L}^p(\Omega,\mathcal{F},P)$, the distance between $X$ and $Y$ is given by
> $$
> d_p(X,Y) := \|X-Y\|_p.
> $$
 
 
 ### Key numerical characteristics of random variables
 
The expected value is a numerical characteristic of a random variable. It represents the long-run average value and is often used as a predictor for the random variable  after a large number of experiments. Note that it does not predict the outcome of a single experiment, but rather the average outcome over a large number of trials. Based on the expected value, we can do some rational decisions very often. 
 
**Example 2.23** A casino has a lottery game where you buy a ticket for $10, and you have a 1 in 100 chance of winning a $500 prize. Otherwise, you get nothing. Should you play? 
 
Let's $X$ the net gain. Then, $X$ can take only two values: $490$ (with probability $0.01$) or $-10$ (with probability $0.99$). The expected value of $X$ is:

$$
\mathbb E[X] = 490 \cdot 0.01 + (-10) \cdot 0.99 = 4.9 - 9.9 = -5.
$$

So you might win big once, but in the long run, the losses overcomes the wins and you will end up losing 5 dollars. So it is not a good idea to play. People that take the chance are giving away their money to the casino. $\blacksquare$ 
 
The previous example shows how the expected value can guide rational choices, but in many cases it is not sufficient by itself. In finance, for instance, two investment opportunities may offer the same expected return, yet a risk-averse investor could still prefer one over the other. A common way to distinguish between them is to measure the typical deviation from the mean. A larger deviation signals greater exposure to randomness, whereas a smaller deviation indicates that outcomes cluster more tightly around the mean, making the distribution more predictable. The following concept is a great tool to measure the typical squared deviation(or dispersion) from the mean.

 
>**Definition 2.23 (Variance)** Let $X$ be a real-valued r.v. The **variance** of $X$ is given by
> $$
> \text{Var}(X) := \mathbb E[(X-\mathbb E[X])^2].
> $$ 

We can easily proof the following properties of the variance:

>**Proposition 2.24** Let $X$ be a real-valued r.v. Then,
>
> 1. $\text{Var}(X) \ge 0$.
> 2. $\text{Var}(X) = 0$ if and only if $X = \mathbb E[X]$ a.s.
> 3. $\text{Var}(aX + b) = a^2 \text{Var}(X)$ for any constants $a$ and $b$.
> 4. $\text{Var}(X) = \mathbb E[X^2] - \mathbb E[X]^2$.
 
>**Definition 2.24 (Standard deviation or volatility)** The **standard deviation** of $X$ is given by
> $$
> \sigma(X) := \sqrt{\text{Var}(X)}.
> $$

The standard deviation is easier to interpret than the variance, since it is on the same scale of the data, and it is the distance between the mean $\mathbb E[X]$ and the data $X$ in $\mathcal{L}^2(\Omega,\mathcal{F},P)$. 
 
 Note that the variance is the second moment of the centered random variable $X-\mathbb E[X]$. In general, the $k$-th moment of a random variable $X$ is given by $\mathbb E[X^k]$. The utility of moments with $k=3,\ 4$ will be clear when we study the properties of the normal distribution, however, we introduce the meaning of those moments in the following definitions: 

>**Definition 2.25 (Skewness)** The **skewness** of $X$ is given by
> $$
> \text{Sk}(X) := \mathbb E\left[\left(\frac{X-\mathbb E[X]}{\sigma(X)}\right)^3\right].
> $$
 
 The skewness is a measure of dierection and degree of asymmetry of the distribution of $X$. The definition performs well measuring the skewness since: 
 
1- Positive deviations contribute positive mass, and negative deviations contribute negative mass, if the distribution is symmetric, the positive and negative contributions cancel out, and the skewness is zero. 
 
2- Cubing the deviations makes the small deviations much smaller than the large deviations, so the $Sk$ is driven by the shape of the tails, not the local slope at the peak. 
 
3- If the right tail is heavier than the left tail, $Sk>0$ and the distribution is said to be **right-skewed** (or left-tailed). If the left tail is heavier than the right tail, $Sk<0$ and the distribution is said to be **left-skewed** (or right-tailed).

4- The $Sk$ is unitless, so it can be used to compare the asymmetry of distributions with different units.

 >**Definition 2.26 (Kurtosis)** The **kurtosis** of $X$ is given by
> $$
> \text{Kt}(X) := \mathbb E\left[\left(\frac{X-\mathbb E[X]}{\sigma(X)}\right)^4\right].
> $$ 

The kurtosis is a measure of the peakedness/tail-heaviness of the distribution of $X$. Tail heaviness is a way of describing how much probability mass a random variable puts in its extreme values (left and right tails). Raising to the fourth power exhagerates the effect of the outliers. This characteristic is compared with benchmark such as the normal distribution.

### Riemann-Stieltjes integral and expected value
 
The reason why from the beginning we defined the expected value as the Lebesgue integral of the random variable is that the Riemann-Stieltjes integral has some deficiencies when dealing with certain wild functions that can be handled well by the Lebesgue integral. However, there are many cases when the Lebesgue integrar agrees with the Riemann-Stieltjes integral. Recall that Riemann integral for a bounded function $f$ on a closed interval $[a,b]$ is defined by using the Riemann sum:

$$
R_n(f,P) = \sum_{i=1}^n f(x_i) (x_i - x_{i-1}),
$$

where $P = \{x_0, x_1, \ldots, x_n\}$ is a partition of $[a,b]$. Then the integral is a number $A$ that satisfies:

$$
\forall \epsilon > 0, \exists \delta > 0: \text{if } P \text{ is a partition of } [a,b] \text{ with } \max_{i=1,\ldots,n} (x_i - x_{i-1}) < \delta, \text{then } |A - R_n(f,P)| < \epsilon.
$$

If we define the norm of the partition $P$ as $\|P\| = \max_{i=1,\ldots,n} (x_i - x_{i-1})$, then the definition of the integral is equivalent to:

$$
\lim_{\|P\|\to 0} R_n(f,P) = A.
$$

If in the above definitions, we consider a real valued function $\gamma$ defined on the interval $(a,\ b]$, and take differences of the form $\gamma(x_i) - \gamma(x_{i-1})$ instead of $x_i - x_{i-1}$, then we get the Riemann-Stieltjes integral: 
 
 $$
 \int_a^b f(x) d\gamma(x) = \lim_{\|P\|\to 0} \sum_{i=1}^n f(x_i) (\gamma(x_i) - \gamma(x_{i-1})).
 $$

So we can get back the Riemann integral by taking $\gamma(x) = x$. Next, our goal is to show how we can go from Lebesgue integral to Riemann-Stieltjes integral when the conditions are met. 
 
 >**Theorem 3.1(Change of variables formula)** Let $X$ be a real-valued random variable, and let $g:\mathbb{R}\rightarrow\mathbb{R}$ be a measurable function respect to $\langle \mathcal{B}(\mathbb{R}), \mathcal{B}(\mathbb{R}) \rangle$. Then,
>
> $$
> \int_{A} g(X) dP = \int_{\mathbb{X}^{-1}(A)} g(x)dP_X = \int_{\mathbb{X}^{-1}(A)} g(x)dF_X,
> $$
>
> where $P_X$ is the distribution of $X$.

**Proof:** You can check the proof in Chapter 2, Section 2.6 of "Probability-1" from Albert N. Shiryaev. $\blacksquare$  
 
Taking $A=\Omega$, we get that

$$
\int_{\Omega} g(X) dP = \int_{\mathbb R} g(x)dF_X,
$$


Note that the above equation is given independently of the distribution of $X$. Often, many of the random variables that we are interested can be integrated using the Riemann-Stieltjes integral as we will see in the following sections, for now, it is enoguh to know that the Riemann-Stieltjes integral when exists, agrees with the Lebesgue integral. The following result summarizes this fact, and we quote it without proof.

>**Theorem 3.4** if the Riemann-Stieltjes integral exists, so does the Lebesgue integral, and the two integrals are equal. 
 
The user can verify that the converse of the above theorem is not true, take for example, $f(x) = 1_{\mathbb I}$, the indicator of irrationals real numbers. However, Lebesgue integrable functions can be well approximated by Riemann integrable functions:

>**Theorem 3.5** Let $f:\mathbb{R}\rightarrow\mathbb{R}$ be a Lebesgue integrable function, then for all $\epsilon > 0$, there exists
>
>i. a step function $g$ such that $\int_{\mathbb{R}} |f(x)-g(x)| dx < \epsilon$,
>
>ii. a continuous integrable function $h$ such that $\int_{\mathbb{R}} |f(x)-h(x)| dx < \epsilon$.






## Discrete random variable distributions
 
 The probability distribution $P_X$ of a discrete random variable has into account that the range of $X$ partitions the sample space $\Omega$ into disjoint events, this is, $\Omega = \bigcup_{x\in A} \{X=x\}$, where $P(X\in A) = 1$, $|A| < \infty$. This leads to the following fact for any event $E\in \mathcal{F}$:
 $$
 P(E) = \sum_{x\in A} P(E\cap X=x) = \sum_{x\in A} P(E|X=x)P(X=x).
 $$
 
 The analytical expression of $P_X$ often is deduced from combinatorial reasoning when the sample space $\Omega$ is finite, otherwise we need to use more sophisticated tools from analysis. The following result gives the analytical expression of the expected value of a discrete random variable.

>**Theorem 2.3** Let $X$ be a discrete random variable, and let $g:\mathbb{R}\rightarrow\mathbb{R}$ be a measurable and integrable function. Then, the expected value of $g(X)$ is given by
>
>$$
>   \mathbb E[g(X)] = \sum_{x\in A} g(x) P(X=x),
>$$
>
>where $A$ is the range of $X$, and it is a countable set. 
 
 **Proof:** If $g(x) = 1_B(x)$ for some $B\in\mathcal{B}(\mathbb{R})$, then by definition of expected value for simple r.v, we have

$$
\mathbb E[1_B(X)] = P(X\in B) = \sum_{x\in A} 1_B(x) P(X=x).
$$ 
 
 If $g(x) = \sum_{i=1}^n a_i 1_{B_i}(x)$ for some $a_i\in\mathbb{R}$ and $B_i\in\mathcal{B}(\mathbb{R})$, then by linearity of the expected value, we have

\begin{align*}
\mathbb E[g(X)] = \sum_{i=1}^n a_i \mathbb E[1_{B_i}] &=\sum_{i=1}^n a_i \sum_{x\in A} 1_{B_i}(x) P(X=x)\\
  &= \sum_{x\in A} \sum_{i=1}^n a_i 1_{B_i}(x) P(X=x) \\
  &= \sum_{x\in A} g(x) P(X=x)
\end{align*}
 
Because the outer sum is finite, exchanging the two sums is unproblematic.For $g$ being a non-negative simple function, the we always can find a sequence of simple non-negative functions $g_n(x)\uparrow g(x)$ for every $x$. Then
\begin{align*}
\mathbb{E}\bigl[g(X)\bigr]
  &= \lim_{m\to\infty} \mathbb{E}\bigl[g_m(X)\bigr] \\
  &= \lim_{m\to\infty} \sum_{x \in A} g_m(x)\,P\!\bigl(X = x\bigr) \\
  &= \sum_{x \in A} g(x)\,P\!\bigl(X = x\bigr).
\end{align*}

The last step is justified by applying MCT intepreting the weighted sum as an integral. For the general case, we write $g = g^{+} - g^{-}$ with $g^{\pm} \ge 0$.
If $\mathbb{E}[\,|g(X)|\,] < \infty$, then both expectations of $g^{\pm}(X)$ are finite, and
$$
\mathbb{E}[g(X)]
   = \mathbb{E}[g^{+}(X)] - \mathbb{E}[g^{-}(X)]
   = \sum_{x \in A} \bigl[g^{+}(x) - g^{-}(x)\bigr] \, P(X = x)
   = \sum_{x \in A} g(x) \, P(X = x).
$$

$\blacksquare$

Taking $g(x) = x$, we get:

>**Corollary 2.4** Let $X$ be a discrete random variable with range $A$, then
>
>$$
>\mathbb E[X] = \sum_{x\in A} x P(X=x).
>$$

The following theorem is a tool for deriving atlernative ways of computing expected value for a nonnegative discrete random variable: 
 
>**Theorem 2.5** Let $X$ be a nonnegative discrete random variable. Then, for nonnegative real values $a_j,\ j\geq 1$, if we define $\displaystyle g(k) = \sum_{j=1}^k a_j$, then
>
>$$
>\mathbb E[g(X)] = \sum_{k=1}^\infty \left(\sum_{j=1}^k a_j\right) P(X=k) = \sum_{j=1}^\infty a_j P(X\geq j)
>$$

**Proof:** Since all the terms in the sum are nonnegative, we can change the order of summation. Also,  
$$\{(j,k): 1\leq j \leq k,\ 1\leq k \lt \infty\} = \{(j,k):  j\leq k \lt \infty ,\ 1\leq j \lt \infty\}$$ 
then we have

\begin{align*}
\sum_{k=1}^\infty \left(\sum_{j=1}^k a_j\right) P(X=k) &= \sum_{j=1}^\infty a_j \left(\sum_{k=j}^\infty P(X=k)\right)\\
&= \sum_{j=1}^\infty a_j P(X\geq j)
\end{align*}



$\blacksquare$


In the above theorem, we can take $a_j = 1$ for all $j\geq 1$, then we get the following result:

>**Corollary 2.6** Let $X$ be a nonnegative discrete random variable. Then,
>
>$$
>\mathbb E[X] = \sum_{j=1}^\infty P(X\geq j)
>$$

**Example 2.7** We want a sample from a set of  $N$ items, where each item is equally likely to appear in each draw. What is the expected number of extractions before seeing a repeated item?

Let $X = \min\{k \ge 2 : \text{at the } k\text{-th purchase you get an item already seen}\} $ be the number of extractions **until the first repetition**. Then $ X $ is a discrete random variable taking values in  $\{2, 3, 4, \ldots\} $.

The event  $\{X > k\}$ means that the first $k$ items are all distinct, so:
$$
P(X > k) = \frac{N (N-1) (N-2) \cdots (N - k + 1)}{N^k} = \prod_{i=1}^{k-1} \left(1 - \frac{i}{N}\right)
$$

Thus, by Corollary 2.6 , we have:
$$
\mathbb{E}[X] = \sum_{k=2}^{\infty} P(X \ge k) = 1 + \sum_{k=3}^{\infty} \prod_{i=1}^{k-2} \left(1 - \frac{i}{N} \right)\approx \sqrt{N\pi/2}
$$

The approximation comes from asymptotic analysis of the product.  

### Examples of discrete random variable distributions
**Bernoulli distribution** 
We say that $X$ has a Bernoulli distribution with parameter $p$ if $P(X=1) = p$ and $P(X=0) = 1-p$. The CDF of $X$ is given by

$$
F_X(x) = 
\begin{cases}
0, & \text{if } x < 0,\\
1-p, & \text{if } 0 \leq x < 1,\\
1, & \text{if } x \geq 1,\\
\end{cases}
$$

Note that Bernoulli distribution can also be stated as $P(X=x) = p^x (1-p)^{1-x}$ for $x \in \{0,1\}$. It is trivial that $P(X=1)+P(X=0)=1$. The experiment that leads to this distribution is the so called **Bernoulli trial**, which is a random experiment with two possible outcomes, usually called success and failure. An example of this experiment is the toss of a coin, where success is defined as "head" and failure as "tail".  The expected value of a Bernoulli random variable and its variance are given by

$$
\mathbb E[X] = 0 \cdot P(X=0) + 1 \cdot P(X=1) = 0 \cdot (1-p) + 1 \cdot p = p\\
\text{Var}(X) = p(1-p)
$$


 
 If we make $n$ independent Bernoulli trials, we may be interested in the probability of getting exactly $k$ successes. To compute this probability, we use combinatorics. There are $\binom{n}{k}$ ways to choose $k$ outcomes that are successes out of $n$ trials, and each success has probability $p$ of occurring, while each failure has probability $1-p$ of occurring. Therefore, the probability of getting exactly $k$ successes is given by the expression in the following definition.

 
**Binomial distribution** 

Let $X$ be the number of successes in $n$ independent Bernoulli trials with success probability $p$. Then, $X$ has a binomial distribution with parameters $n$ and $p$, denoted by $X \sim \text{Binomial}(n,p)$. The distribution of $X$ is given by

$$
P(X=k) = \binom{n}{k} p^k (1-p)^{n-k}
$$ 

It is not difficult to show that the sum of the probabilities of all possible values of $X$ is 1, i.e., $\sum_{k=0}^n P(X=k) = 1$. This is a consequence of the binomial theorem applied to the expansion of $(p+(1-p))^n$.
 
 **Example 2.5** 7 coins are tossed independently, and the number of heads is recorded. What is the probability of getting exactly $3$ heads if every coin has probability $p=3/5$ of landing heads? 

Let $X$ be the number of heads in the $7$ tosses. Then, $X$ has a binomial distribution with parameters $n=7$ and $p=3/5$, i.e., $X \sim \text{Binomial}(7,3/5)$. The probability of getting exactly $3$ heads is given by

$$
P(X=3) = \binom{7}{3} (3/5)^3 (2/5)^{7-3} = 0.1166
$$

 $\blacksquare$ 

 The CDF of the binomial distribution is given by
 
 $$
 F_X(x) = \sum_{k \in \mathbb{N}_0:k\leq x} \binom{n}{k} p^k (1-p)^{n-k}
 $$
  
By checking the behavior of the distribution as variable of $k$, we can stablish some interesting facts. Basic computation shows that 
$$
\frac{P(X=k+1)}{P(X=k)} = \frac{n-k}{k+1} \frac{p}{q}
$$

where $q=1-p$. Comparing this ratio with $1$, we see that 
 
i. $P(X=k+1) > P(X=k)$ if and only if $k<np-q$.

ii. $P(X=k+1) < P(X=k)$ if and only if $k > np-q$.
 
iii. $P(X=k+1) = P(X=k)$ if and only if $k = \lfloor np-q \rfloor$. If $np-q$ is an integer, the equality shows that the maximum is attained at two points, namely $k=np-q$ and $k'=np-q+1$.

Therefore, the distribution has its maximum at $k = \lfloor np-q \rfloor$(and $k'=np-q+1$ if $np-q$ is an integer). Easy computation shows that the expected value and variance of a binomial random variable are given by

$$
\mathbb E[X] = np\\
\text{Var}(X) = np(1-p)
$$ 

In fact, we can derive a recursive formula for the expected value and variance of a binomial random variable, by applying the change $k = j-1$ in the definition of $\mathbb E[X^m]$, so the next we get the next result: 

>**Proposition 2.6** Let $X\sim \text{Binomial}(n,p)$, then
>
>$$
>\mathbb E[X^m] = np \mathbb E[(Y+1)^{m-1}]
>$$
>
>where $Y\sim \text{Binomial}(n-1,p)$. 
 
**Proof:** Left as an exercise. $\blacksquare$




We may encounter situations where the number of trial is large, and the computation of the binomial coefficient may be cumbersome.  Fortunately, there are ways to approximate the binomial distribution. To this end, there are two predominant cases regarding the value of $p$:

i. $p$ is fixed independent of $n$. In this case, the distribution is symmetric and the binomial distribution can be approximated by the normal distribution as we'll see later.

ii. $p$ is close to $0$ for big $n$. In this case, the distribution is skewed and the binomial distribution can be approximated by the Poisson distribution as we describe when we introduce the Poisson distribution next. 
 
**Poisson distribution**

In the context of the binomial distribution, when the occurrence of certain event is rare for a huge number of trials, the Poisson distribution is a suitable approximation. 
 
 >**Theorem 2.7** Let $X\sim \text{Binomial}(n,p_n)$, and let $\lambda_n = np_n$. If $\lim_{n\to\infty} \lambda_n = \lambda \in (0,\infty)$, then
 >
 >$$
 > \lim_{n\to\infty} P(X=k) = \frac{\lambda^k}{k!} e^{-\lambda}
 >$$

 **Proof:** From definition of binomial coefficient and basic results from real analysis, we have: 
  
  1. $\lim_{n\to\infty} \binom{n}{k}/n^k = \frac{1}{k!}$

  2. $\lim_{n\to\infty} \left(1+\frac{x_n}{n}\right)^n = e^{\lim_{n\to\infty} x_n}$.  
  
 Then, from the definition of binomial distribution, we have: 
 
 \begin{align*}
 P(X=k) &= \binom{n}{k} p_n^k (1-p_n)^{n-k} = \binom{n}{k} \left(\frac{\lambda_n}{n}\right)^k \left(1-\frac{\lambda_n}{n}\right)^{n-k}\\
 &= \bigg(\binom{n}{k}/n^k\bigg) \lambda_{n}^k \left(1-\frac{\lambda_n}{n}\right)^{n-k}\\
 &= \bigg(\binom{n}{k}/n^k\bigg) \lambda_{n}^k \frac{\left(1-\frac{\lambda_n}{n} \right)^{n}}{\left(1-\frac{\lambda_n}{n}\right)^k}\\
 \end{align*}
 
 Now, we can apply the limit to both sides of the equation, 
 
 \begin{align*}
 \lim_{n\to\infty} P(X=k) &= \lim_{n\to\infty} \bigg(\binom{n}{k}/n^k\bigg) \lambda_{n}^k \frac{\left(1-\frac{\lambda_n}{n} \right)^{n}}{\left(1-\frac{\lambda_n}{n}\right)^k}\\
 &= \frac{\lambda^k}{k!} e^{-\lambda}
 \end{align*} 
  
$\blacksquare$

>**Definition 2.6 (Poisson distribution)** A random variable $X$ has a Poisson distribution with parameter $\lambda > 0$ if its distribution is given by
>
>$$
>P(X=k) = \frac{\lambda^k}{k!} e^{-\lambda}
>$$


The reader can easily verify that $\sum_{k=0}^\infty P(X=k) = 1$, so the distribution is a valid probability distribution.

Note that in the process of proving the theorem, we took $\lambda = np$, which can be considered as the average(or expected) number of occurrences of the event in the $n$ trials for a binomial distribution. 
 
**Example 2.6** A bank has 900 borrowers, and the probability that a borrower will default is $p=0.01$. What is the probability that exactly 7 borrowers will default?

Let $X$ be the number of borrowers that will default. Then, $X$ has a binomial distribution with parameters $n=900$ and $p=0.01$. The probability that exactly 7 borrowers will default is given by

$$
P(X=7) = \binom{900}{7} (0.01)^7 (0.99)^{900-7} \approx 0.11731
$$

but this is a very difficult computation, so we can use the Poisson distribution to approximate the binomial distribution. The average number of defaults is $\lambda = np = 900 \times 0.01 = 9$. Therefore, the probability that exactly 7 borrowers will default is given by

$$
P(X=7) = \frac{9^7}{7!} e^{-9} \approx 0.11712
$$

We see that the Poisson approximation error is less than $2\%$, and the computation is much easier.

$\blacksquare$

**Example 2.7** A particular stock has an average of 60 trades per hour. If a trader steps away from the trading desk for 30 seconds, what is the probability that there will be no trades during that interval?

Let $X$ be the number of trades in the interval of 30 seconds. Since $1h=3600s = 120\times 30s$, then there are 120 intervals of 30 seconds in an hour. The average number of trades in an hour is 60, so the average number of trades in 30 seconds is $\lambda = 60/120 = 0.5$. Therefore, $X\sim \text{Poisson}(0.5)$. The probability that there will be no trades during that interval is given by

$$
P(X=0) = \frac{0.5^0}{0!} e^{-0.5} = e^{-0.5} \approx 0.6065
$$

$\blacksquare$

The characteristics of Poisson distribution are given by:

$$
\mathbb E[X] = \lambda\\
\text{Var}(X) = \lambda
$$

Since the Poisson distribution is a limit of the binomial distribution, it is also a valid the following result:

>**Proposition 2.8** Let $X\sim \text{Poisson}(\lambda)$, then
>
>$$
>\mathbb E[X^m] = \lambda \mathbb E[(Y+1)^{m-1}]
>$$
>
>where $Y\sim \text{Poisson}(\lambda)$. 
 
**Proof:** Left as an exercise. $\blacksquare$
 
**Geometric distribution**

Consider a sequence of independent Bernoulli trials with success probability $p$. We want to know the probability that the first success occurs at the $k$-th trial. The answer is given by the following definition.

>**Definition 2.7 (Geometric distribution)** A random variable $X$ has a geometric distribution with parameter $p$, denoted by $X\sim \text{Geometric}(p)$, if its distribution is given by
>
>$$
>P(X=k) = (1-p)^{k-1} p
>$$ 
 
The above equation follows from the fact that the probability of the first success at the $k$-th trial is the probability of $k-1$ failures followed by a success. Using the definition of geometric series, we can check that $\sum_{k=1}^\infty P(X=k) = 1$, so the distribution is a valid probability distribution.  
 
**Example 2.8** In a simple trading algorithm, the trade can make money with estimated probability $p=0.4$. If the trader keeps running the algorithm until the first loss appears, what is the probability that the first loss appears at the 5th trade?

Let $X$ be the number of trades until the first loss appears. Then, $X\sim \text{Geometric}(0.4)$, so the required probability is given by

$$
P(X=5) = (1-0.4)^{5-1} \times 0.4 = 0.6^4 \times 0.4 = 0.05184
$$

$\blacksquare$


The expected value and variance of the geometric distribution are given by:

$$
\mathbb E[X] = \frac{1}{p}\text{Var}(X) = \frac{1-p}{p^2}
$$ 
 
 For the computation of the expected values, use term by term derivative of the power series $f(x) = \sum_{k=1}^\infty kx^{k-1}$, and evaluate at $x=1-p$.

##  Absolutely continuous random variables

A special case of continuous random variables is the absolutely continuous random variables, which appear in many models of real-world phenomena.

>**Definition 3.2 (Absolutely continuous random variable)** A random variable $X$ is said to be absolutely continuous if there exists a non-negative Lebesgue integrable function $f_X:\mathbb{R}\rightarrow\mathbb{R}$ such that for all $B\in\mathcal{B}(\mathbb{R})$,
>
>$$
>P(X\in B) = \int_B f_X(x) dx.
>$$
>If we take $B = (-\infty, x]$, we get the cumulative distribution function of $X$ as
>
>$$
>F_X(x) = \int_{-\infty}^x f_X(t) dt.
>$$


In the above definition, the integral is respect to Lebesgue measure, but to avoid technicalities, we assume the integrability in the Riemann sense. Since $\lim_{x\to\infty} F_X(x) = 1$, then $\int_{-\infty}^\infty f_X(x) dx = 1$. The function $f_X$ is called the **probability density function of $X$**.

**Example 3.3** Let $X$ be a random variable with a uniform distribution on the interval $(0,1)$ as in the example 2.6. Then, $X$ is absolutely continuous with density function $f_X(x) = 1_{(0,1)}(x)$, since $F_X(x) = x = \int_{-\infty}^x 1_{(0,1)}(t) dt$.

Similar to discrete case, we can compute the probability of an event by integrating the  product of the conditional probability and the density function. This is: 

$$
P(X\in B) = \int_B P(X\in B|X=x) f_X(x) dx.
$$

To prove this, we need results like Radon-Nikodym theorem, which we will see when it is needed. The way of computing the expected value of an absolutely continuous random variable is given by the following theorem:

>**Theorem 3.4** Let $X$ be an absolutely continuous random variable with density function $f_X(x)$, and let $g:\mathbb{R}\rightarrow\mathbb{R}$ be a Borel measurable function such that $\mathbb{E}[|g(X)|] < \infty$. Then, the expected value of $g(X)$ is given by
>
>$$
>\mathbb E[g(X)] = \int_{-\infty}^\infty g(x) f_X(x) dx.
>$$

**Proof:** If $g(x)=1_B(x)$ for some $B\in\mathcal B(\mathbb R)$, then using change of variables, we have
$$
\mathbb E[g(X)] = P(X\in B) = \int_B f_X(x) dx = \int_{-\infty}^\infty 1_B(x) f_X(x) dx = \int_{-\infty}^\infty g(x) dF_X(x).
$$
 
 If $g(x)=\sum_{i=1}^n a_i 1_{A_i}(x)$ for some $a_i\in\mathbb{R}$ and $A_i\in\mathcal B(\mathbb R)$, where $A_i$ are disjoint sets (we can always find such a representation, why?), then
\begin{align*}
\mathbb E[g(X)] = \sum_{i=1}^n a_i \mathbb E[1_{A_i}(X)] &= \sum_{i=1}^n a_i P(X\in A_i) \\
&= \sum_{i=1}^n a_i \int_{A_i} f_X(x) dx \\
&= \sum_{i=1}^n a_i \int_{-\infty}^\infty 1_{A_i}(x) f_X(x) dx \\
&= \int_{-\infty}^\infty \sum_{i=1}^n a_i 1_{A_i}(x) f_X(x) dx \\
&= \int_{-\infty}^\infty g(x) f_X(x) dx.
\end{align*}

If $g(x)$ is a non-negative Borel measurable function, then $g(X)$ is a random variable, and we can find a sequence of non-negative simple functions $g_n(x)$ such that $g_n(x)\uparrow g(x)$ for all $x\in\mathbb{R}$. Using the previous step, we have that $\mathbb E[g_n(X)]=\int_{-\infty}^\infty g_n(x) f_X(x) dx$. Then, by the monotone convergence theorem, we have that

$$
\mathbb E[g(X)] = \lim_{n\to\infty} \mathbb E[g_n(X)] = \int_{-\infty}^\infty \lim_{n\to\infty} g_n(x) f_X(x) dx = \int_{-\infty}^\infty g(x) f_X(x) dx.
$$ 
 
For the case of a general Borel measurable function $g(x)$, we can write $g(x) = g^+(x) - g^-(x)$. Since $g^+(x)$ and $g^-(x)$ are non-negative Borel measurable functions, we can apply the previous step to each of them. Then, we have that
$$
\mathbb E[g(X)] = \mathbb E[g^+(X)] - \mathbb E[g^-(X)] = \int_{-\infty}^\infty g^+(x) f_X(x) dx - \int_{-\infty}^\infty g^-(x) f_X(x) dx = \int_{-\infty}^\infty g(x) f_X(x) dx.
$$
$\blacksquare$

If we take $g(x)=x$, we get the following result:

>**Corollary 3.5** Let $X$ be an absolutely continuous random variable with density function $f_X(x)$, then
>
>$$
>\mathbb E[X] = \int_{-\infty}^\infty x f_X(x) dx.
>$$

Similarly to the discrete case, we have a result that allows us to derive another ways of computing the expected value for a non-negative absolutely continuous random variable: 
 
>**Theorem 3.6** Let $X$ be a non-negative absolutely continuous random variable with density function $f_X(x)$, and let $h:\mathbb{R}\rightarrow\mathbb{R}$ be a non-negative measurable function, then
>
>$$
>\mathbb E\bigg[\int_{0}^Xh(t)dt\bigg] = \int_{0}^\infty h(t)P(X>t)dt.
>$$

**Proof:** Since $g(x,t) = f_X(x)h(t)\ge 0$ for all $x,t\in\mathbb{R}$, then:

\begin{align*}
\mathbb E\bigg[\int_{0}^Xh(t)dt\bigg] &= \int_{0}^\infty \int_{0}^x h(t)f_X(x)dtdx \\
&= \int_{0}^\infty \int_{t}^\infty h(t)f_X(x)dxdt\quad \text{by Fubini's theorem}\\
&= \int_{0}^\infty h(t) \int_{t}^\infty f_X(x)dx dt \\
&= \int_{0}^\infty h(t)P(X>t)dt.
\end{align*}

$\blacksquare$

If we take $h(t) = 1$, we get the following result:

>**Corollary 3.7** Let $X$ be a non-negative absolutely continuous random variable with density function $f_X(x)$, then
>
>$$
>\mathbb E[X] = \int_{0}^\infty P(X>t)dt.
>$$

**Uniform distribution**

>**Definition 3.5 (Uniform distribution)** A random variable $X$ has a uniform distribution on the interval $(a,b)$ if its density function is given by
>
>$$
>f_X(x) = \frac{1}{b-a} 1_{(a,b)}(x).
>$$ 
>
>So, the CDF of the uniform distribution is given by
>
>\begin{align*}
>F_X(x) &= \begin{cases}
>0, & x \leq a, \\
>\frac{x-a}{b-a}, & a < x < b, \\
>1, & x \geq b.
>\end{cases}
>\end{align*}



**Example 3.9** A man arrives to a bus stop at most one hour after noon. if buses arrive at ten minute intervals starting at noon, what is the probability that the man will have to wait less than 5 minutes for the next bus? Assume that the man arrives at the bus stop at a random time within the hour.

Let $X$ be the time the man arrives at the bus stop. Then, $X$ has a uniform distribution on the interval $(0,60)$. The probability that the man will have to wait less than 5 minutes for the next bus is given by

$$
P\left(\bigcup_{k=1}^6 P(10k-5 < X < 10k)\right) = \sum_{k=1}^6 P(10k-5 < X < 10k) = \sum_{k=1}^6 \frac{5}{60} = \frac{1}{2}.
$$
$\blacksquare$


The expected value and variance of the uniform distribution are given by:

$$
\mathbb E[X] = \frac{a+b}{2}\quad \text{and}\quad \text{Var}(X) = \frac{(b-a)^2}{12}.
$$

The uniform distribution is very useful in sampling according to a given distribution. To this end, we can use the inverse transform sampling method.
 
>**Theorem 3.10** Let $F$ a distribution function, and let $U$ be a random variable uniformly distributed on the interval $[0,1]$. Then,
>
>i. If $F$ is continuous, then $X = F^{-1}(U)$ has distribution function $F$.
>
>ii. If $F$ is the CDF of r.v taking non-negative values in $\mathbb Z$, then the r.v 
>$$
>X = \min\{k\in\mathbb Z: F(k) \ge U, \ k\ge 0\}
>$$
>has distribution function $F$.

**Proof:** i. Let $x\in\mathbb{R}$. Then, because $F$ is continuous and nondecreasing:

$$
P(X \leq x) = P(F^{-1}(U) \leq x) = P(U \leq F(x)) = F(x).
$$

ii. Let $k\in\mathbb{Z}, \ k\ge 0$. Then, because $F$ is the CDF of r.v taking non-negative values in $\mathbb Z$, we have that $F(k) \in [0,1]$. Then,

$$
P(X \le x) = \sum_{k=0}^{\lfloor x \rfloor} P(X=k) = \sum_{k=0}^{\lfloor x \rfloor} (F(k+1) - F(k)) = F(x).
$$

since $F(\lfloor x \rfloor) = F(x)$ when $X$ takes non-negative integer values (why?). $\blacksquare$ 

The steps to generate a random variable $X$ with a given distribution function $F$ are the following:

1. Generate a random number $U$ from the uniform distribution on the interval $[0,1]$.
2. Compute $X = F^{-1}(U)$ or the minimum $k\in\mathbb{Z}$ such that $F(k) \ge U$.

The following example shows how to generate a binomial random variable using the inverse transform sampling method.
 
**Example 3.11** Let $U$ an uniform random variable on the interval $[0,1]$, and let $F(x)$ be the CDF of a random variable $X\sim \text{Binomial}(n,p)$. The following script shows how to generate a binomial random variable using the inverse transform sampling method.

**Exponential distribution**

Suppose that we are observing certain events occurring at random times. For instance, we can think of the events as phone calls arriving at a call center, or cars passing a certain point on a road. We ware interested in the number of events $N_t$ that occur in a given interval of time $t$. A standard way to model this is through the **Poisson process**, which is a stochastic process that satisfies the following properties:

1. $N_t\ge 0$, and $N_0=0$.
2. The number of events in disjoint intervals are independent.
3. The distribution of the number of events in an interval depends only on the length of the interval and not on its location.
4. The probability of exactly one event occurring in an interval of length $h$ is $P(N_{t+h} - N_t = 1) = \lambda h + o(h)$, for some constant $\lambda > 0$, which is called the **rate** of the process.
5. The probability of more than one event occurring in an interval of length $h$ is $P(N_{t+h} - N_t > 1) = o(h)$.

Then we have the following result:

>**Theorem 3.12** Let $N_t$ be a Poisson process with rate $\lambda > 0$. Then, the distribution of the time $T_1$ until the first event occurs is given by
>
>$$
>P(T_1 > t) = e^{-\lambda t}.
>$$ 
 
**Proof:** First, we derive the form of $P(N_t = 0)$. We have that: 
\begin{align*}
P(N_{t+h} = 0) &= P(N_t = 0,\ N_{t+h} - N_t = 0) \\
&= P(N_t = 0)P(N_{t+h} - N_t = 0) \quad \text{by independence} \\
&= P(N_t = 0)(1 - \lambda h + o(h)) \quad \text{by property 4} \\
\end{align*}

After breaking the parenthesis, we can transform the expression to get:

$$
\frac{P(N_{t+h} = 0) - P(N_t = 0)}{h} = - \lambda P(N_t = 0) + \frac{o(h)}{h}
$$ 

Taking the limit as $h$ approaches 0, we get:

$$
\lim_{h\to 0} \frac{P(N_{t+h} = 0) - P(N_t = 0)}{h} = \frac{dP(N_t = 0)}{dt} = - \lambda P(N_t = 0)
$$

Solving the differential equation with the initial condition $P(N_0 = 0) = 1$, we get:

$$
P(N_t = 0) = e^{-\lambda t}
$$

This means that the probability of no events occurring in the interval $[0,t]$ is $e^{-\lambda t}$. Therefore, the probability of the first arrival $T_1$ after $t$ is given by:

$$
P(T_1 > t) = P(N_t = 0) = e^{-\lambda t}
$$

$\blacksquare$

The distribution of $T_1$ is called the **exponential distribution**, as we formalize in the following definition.

>**Definition 3.6 (Exponential distribution)** A random variable $X$ has an exponential distribution with parameter $\lambda > 0$ if its density function is given by
>
>$$
>f_X(x) = \lambda e^{-\lambda x} 1_{(0,\infty)}(x).
>$$
> 
> The CDF is:
> 
>\begin{align*}
>F_X(x) &= \begin{cases}
>0, & x \leq 0, \\
>1 - e^{-\lambda x}, & x > 0.
>\end{cases}
>\end{align*}
>

The exponential distribution arises naturally as the waiting time until the first arrival in a Poisson process. In practical applications, it models the time between independent, randomly occurring events at a constant average rate $$, such as the time until the next phone call, customer arrival, or default event. 

When computing the expected value, if we do integration by parts, we get:

$$
\mathbb E[X^n] = \frac{n}{\lambda} \mathbb E[X^{n-1}]
$$

Using this result, we can see that the expected value and variance of the exponential distribution are given by:

$$
\mathbb E[X] = \frac{1}{\lambda} \quad \text{and} \quad \text{Var}(X) = \frac{1}{\lambda^2}
$$


An important property of the exponential distribution is that it is **memoryless**. This means that the probability of the next event occurring in a given interval of time is independent of the time that has already elapsed. 
 
**Definition 3.7 (Memoryless property)** A random variable $X$ is said to be memoryless if for all $t,s\ge 0$,

$$
P(X > t+s | X > t) = P(X > s).
$$

**Example 3.10** At a brokerage firm, the time required to execute a clients trade request (e.g. placing a stock or option order) follows an exponential distribution with a mean of $a$ minutes. You walk in (or check the dashboard) and see a trade already being processed for more than $b$ minutes. Whats the probability that the execution will still be going on after another $c$ minutes? 
 
Let $X$ be the time required to execute a clients trade request. Then, $X$ follows an exponential distribution with parameter $\lambda = \frac{1}{a}$. The probability that the execution will still be going on after another $b$ minutes is given by:

$$
P(X > c + b | X > b) = P(X > c) = e^{-\lambda c} = e^{-\frac{c}{a}}
$$

$\blacksquare$ 
 
 Next we briefly expose the relationship between the exponential distribution and the hazard rate. If $X$ is the lifetime of a device, and assume that $X$ is absolutely continuous, then a quantity of interest is the instantaneous probability that the device will fail at time $t$ given that it has survived until time $t$. In other words, we are interested in the conditional probability $P(X \le t + dt | X > t)$. So we have that:

$$
P(X \le t + dt | X > t) = \frac{P(X \le t + dt, X > t)}{P(X > t)} = \frac{P(t < X \le t + dt)}{P(X > t)} = \frac{F_X(t + dt) - F_X(t)}{1 - F_X(t)}
$$

Dividing both sides by $dt$ and taking the limit as $dt$ approaches 0, we get:

$$
\lim_{dt \to 0} \frac{P(X \le t + dt | X > t)}{dt} = \frac{f_X(t)}{1 - F_X(t)}
$$

**Definition 3.8 (Hazard rate)** The hazard rate of a random variable $X$ is defined as:

$$
\lambda(t) = \frac{f_X(t)}{1 - F_X(t)}
$$

Note that if $X$ has an exponential distribution with parameter $\lambda$, then $\lambda(t) = \lambda$, this is, the hazard rate is constant. Often, we can consider that the hazard rate is not constant, due to natural factors that make the device more likely to fail as it ages.  If we treat the hazard rate $\lambda(t)$ as known, we can recover the CDF $F_X(t)$ as follows:

$$
\lambda(t) = \frac{f_X(t)}{1 - F_X(t)} \implies \int_0^t \lambda(s) ds = \int_0^t \frac{1}{1 - F_X(s)} dF_X(s) = -\ln(1 - F_X(t))
$$

since $F_X(0) = 0$. Then, taking the exponential of both sides, we get:

$$
F_X(t) = 1-e^{-\int_0^t \lambda(s) ds}
$$

Note that for $\lambda(t) = \lambda$, we get back the exponential distribution. In general, each hazard rate determines a unique distribution. 
 


**Normal distribution and Central Limit Theorem**
 
 
In many real-life scenarios, the overall effect of multiple independent factors can be approximated by the sum of  i.i.d. random variables, which we call **observations**. Examples include the amount of rainfall in a given area over a period, or the number of defects in a manufactured product. These situations often involve numerous small, independent events that contribute to a larger overall outcome. It turns out that the larger the number of observations, the sume of observations becomes less skewed. To show this, let $X_1, X_2, \ldots, X_n$ be i.i.d. random variables with mean $\mu$ and variance $\sigma^2$. Then, the sum $S_n = X_1 + X_2 + \cdots + X_n$ can be written as:

$$
S_n = n\mu + \sum_{i=1}^n Y_i = n\mu + T_n
$$

where $Y_i = X_i - \mu$ and $T_n = \sum_{i=1}^n Y_i$. Now, let's compute the skewness of $S_n$. We have that (why?):

$$
\text{Sk}(S_n) =\text{Sk}(S_n-n\mu) = \text{Sk}(T_n)
$$ 
 
On one hand, $\sigma_{T_n}^2 = Var(T_n) = n\sigma^2$. On the other hand, $\mathbb E[T_n^3] = \mathbb E[\sum_{i=1}^n Y_i^3] = \sum_{i=1}^n \mathbb E[Y_i^3]$ since by independence, the mixed terms are zero. Then, we have that:

$$
\text{Sk}(S_n) = \frac{\mathbb E[T_n^3]}{\sigma_{T_n}^3} = \frac{\mathbb E[\sum_{i=1}^n Y_i^3]}{n^{3/2}\sigma^3}  = \frac{n\mathbb E[Y_i^3]}{n^{3/2}\sigma^3} = \frac{\mathbb E[Y_i^3]}{\sqrt{n}\sigma} = \frac{\text{Sk}(Y_1)}{\sqrt{n}}
$$

Hence, $\text{Sk}(S_n)\to 0$ as $n\to\infty$. This means that the distribution of $S_n$ becomes more and more symmetric as the number of observations increases. The symmetric distribution to which the scaled $S_n$ distribution converges is the **normal distribution**, which we define next. 
 
>**Definition 3.9 (Normal distribution)** A random variable $X$ has a normal distribution with mean $\mu$ and variance $\sigma^2$, written as $X\sim N(\mu, \sigma^2)$, if its density function is given by:
>
>$$
>f_X(x) = \frac{1}{\sqrt{2\pi\sigma^2}} e^{-\frac{(x-\mu)^2}{2\sigma^2}}
>$$ 
 
The user can check that the integral of the density function is 1, either by complex integration or by using multivariate calculus. The convergence of the distribution of $S_n$ is guaranteed by the **Central Limit Theorem**, which we state next. But before that, let's define **convergence in distribution**.

>**Definition 3.10 (Convergence in distribution)** A sequence of random variables $X_1, X_2, \ldots$ converges in distribution to a random variable $X$ if:
>
>$$
>P(X_n \le x) \to P(X \le x) \quad \text{as } n\to\infty
>$$ 
>and we write $X_n \stackrel{d}{\to} X$ as $n\to\infty$. Where $x$ is a continuity point of $F_X$ and $F_{X_n}$.
 
In other words, $F_{X_n}(x) \to F_X(x)$ as $n\to\infty$, for all $x\in\mathbb{R}$.

>**Theorem 3.13 (Central Limit Theorem)** Let $X_1, X_2, \ldots, X_n$ be i.i.d. random variables with mean $\mu$ and variance $\sigma^2$. Let's $S_n = X_1 + X_2 + \cdots + X_n$. Then, 
>
>$$\frac{S_n - n\mu}{\sigma\sqrt{n}} \stackrel{d}{\to} N(0,1) \quad \text{as } n\to\infty.$$


The following script shows how the scaled sum of i.i.d. random variables has bell-shaped distribution:



```python
import numpy as np, matplotlib.pyplot as plt, scipy.stats as st
n, reps = 30, 50_00
samples = np.random.randint(1, 7, size=(reps, n))  # fair dice, highly non-Gaussian
sums = samples.sum(axis=1)
normed = (sums - sums.mean()) / sums.std()

plt.hist(normed, bins=60, density=True, alpha=.6, label='Simulated sums')
x = np.linspace(-4, 4, 400); plt.plot(x, st.norm.pdf(x), 'r', lw=2, label='N(0,1)')
plt.title(f"Sum of {n} fair dice  Normal by CLT"); plt.legend(); plt.show()
```

An important result is the following:

>**Proposition 3.14** Let $X\sim N(\mu, \sigma^2)$. Then, for any $a,b\in\mathbb{R}$ with $a\not = 0$, the random variable $Y = aX + b$ has a normal distribution with mean $a\mu + b$ and variance $a^2\sigma^2$. 
 
**Proof:** Let $Y=aX+b$. Then, the CDF of $Y$ is given by:

$$
F_Y(y) = P(Y \le y) = P(aX + b \le y) = P\left(X \le \frac{y-b}{a}\right) = F_X\left(\frac{y-b}{a}\right)
$$ 

After taking the derivative of $F_y$, we can check the claim. The reader should be able to add the details. $\blacksquare$. 
 

The case when $a=1/\sigma$ and $b=-\mu/\sigma$ correspond to the scaling of $X$, such r.v is called the **standard normal** distribution, and is defined as 
 
 $$
 Z = \frac{X-\mu}{\sigma}
 $$ 

 The CDF of $Z\sim N(0,1)$ is: 

$$
\Phi(z) = \frac{1}{\sqrt{2\pi}}\int_{-\infty}^{z} e^{-\frac{x^2}{2}}
$$


Although there is no closed form of the CDF $\Phi(z)$, there are ways of approximating it with extreme precision and without difficulty. And this is an important fact since all calculations of the CDF of $X\sim N(\mu, \sigma^2)$ can be reduced to calculations for the standard normal distribution since 
$$
P(X\le x) = P\left(\frac{X-\mu}{\sigma}\le \frac{x-\mu}{\sigma}\right) = \Phi\left(\frac{x-\mu}{\sigma}\right)
$$

The following property is easy to prove:

>**Proposition 3.15** Let $Z\sim N(0,1)$. Then, for any $z\in\mathbb{R}$,
>
>$$
>\Phi(-z) = 1-\Phi(z)
>$$ 
 
That is really useful, since having a table of the values of $\Phi(z)$ for $z\ge 0$ is enough to compute the values of $\Phi(z)$ for any $z\in\mathbb{R}$. Next, we list additional notable facts about the normal distribution. 
 
>**Proposition 3.16** Let $Z\sim N(0,1)$. Then, for any $x\in\mathbb{R}$,
>
>1- $68\%$ of the area under the curve is between $-1$ and $1$ (i.e, one standard deviation from the mean). 
>
>2- $95\%$ of the area under the curve is between $-2$ and $2$ (i.e, two standard deviations from the mean).
>
>3- $99\%$ of the area under the curve is between $-3$ and $3$ (i.e, three standard deviations from the mean). 
> The moment of order $k$ of $Z$ is given by:
>$$
>\mathbb E[Z^k] = \begin{cases}
>0, & \text{if } k \text{ is odd}, \\
> 1\cdot 3\cdot 5\cdots (k-1), & \text{if } k \text{ is even}.
>\end{cases}
>$$
> $\text{Sk}(Z) = 0$ and $\text{Kur}(Z) = 3$. 
 
For using the normal distributions as baseline for kurtosis, from now on, we will use the following definition:

>**Definition 3.17 (Kurtosis)** The kurtosis of a random variable $X$ is defined as:
>
>$$
>\gamma_4(X) = \frac{\mathbb E[(X-\mu)^4]}{\sigma^4} - 3
>$$ 
 
So, for $Z\sim N(0,1)$, we have that $\gamma_4(Z) = 0$. 


**Example 3.17(Binomial approximation to normal)** Suppose we flip a coin $n$ times. Knowing the probability of heads is $p>>0.1$, what is the probability of at most $k$ heads? 
 
The CLT applied to the binomial is also known as the **De Moivre-Laplace theorem**. And we use it when there is a big number of trials and $p$ is not too close to $0$ (recall that for $p<<0.1$, the binomial distribution is well approximated by the Poisson distribution). Then, since the number of head $X=\sum_{i=1}^{n} X_i$ is the sum of $n$ i.i.d. Bernoulli random variables, we have by the CLT that:

$$
\frac{X - n\cdot p}{\sqrt{n\cdot p\cdot (1-p)}} \stackrel{d}{\to} N(0,1)
$$

This means that: 
$$
P(X\le k) = P\left(\frac{X - n\cdot p}{\sqrt{n\cdot p\cdot (1-p)}} \le \frac{k - n\cdot p}{\sqrt{n\cdot p\cdot (1-p)}}\right) \approx \Phi\left(\frac{k - n\cdot p}{\sqrt{n\cdot p\cdot (1-p)}}\right)
$$ 
 
 $\blacksquare$
