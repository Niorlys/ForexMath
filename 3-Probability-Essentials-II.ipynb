# Random Variables 
 
When performing experiments, often we are interested in a quantitative characterization of the outcomes instead of the outcomes themselves. For example, when we toss two dice, we might be interested in the sum of the numbers that come up. In this case, we can build a function $X$ where $X(\omega) = X((i,j)) = i+j$. Then, for instance, the event $A = \{(1,2), (2,1)\}$ can be stated as $\{ \omega: X(\omega) = 3\} = X^{-1}(\{3\})$. Another example, when someone is betting on a coin toss, earning the amount bet if the coin lands heads and losing double the amount bet if the coin lands tails. In this case, for an amount $n$, the net return can be stated as $X_n = n I_H - 2n I_T = n(3I_H -2)$, where $I_H$ is the indicator function of the event where the coin lands heads and $I_T$ is the indicator function of the event where the coin lands tails, and $I_T = 1 - I_H$. Informally, we say that a **random variable** is any real-valued function on the sample space $\Omega$. For assigning probabilities to the events mapped by $X$, we must allow $X$ to satisfy the "natural" condition specified by the following definition.  

**Definition 2.1 (Measurable function)** Let $(\Omega, \mathcal{F})$ be a measurable space. A function $f: \Omega \to \mathbb{R^p}$ is  $\langle \mathcal{F}, \mathcal{B}(\mathbb{R^p}) \rangle$ measurable, if for every Borel set $B \subset \mathbb{R^p}$, we have
$$f^{-1}(B) = \{ \omega \in \Omega: f(\omega) \in B \} \in \mathcal{F}.$$  
 
Often, for simplicity of notation, we will write $\{ \omega \in \Omega: f(\omega) \in B \}$ as $\{f \in B\}$.
 
**Definition 2.2 (Random Variable)**  Let $(\Omega, \mathcal{F})$ be a measurable space. A function $X: \Omega \to \mathbb{R^p}$ is called a random variable if it is $\langle \mathcal{F}, \mathcal{B}(\mathbb{R^p}) \rangle$ measurable. This is; for every Borel set $B \in \mathcal{B}(\mathbb{R^p})$, we have 
$$X^{-1}(B) = \{ \omega \in \Omega: X(\omega) \in B \} \in \mathcal{F}$$

  
For simplicity, we take $p=1$ in the definition above, and when there is no ambiguity, we say only that $X$ is a random variable, assuming that preimages are taken with respect to the Borel $\sigma$-algebra in $\mathbb{R^p}$. We call the condition "natural" because, as we saw earlier, events defined through a random variable are described by the preimage of Borel sets. Since we assign a probability measure to these events, it is natural to require that the preimages of Borel sets be measurable. An inmediate example of randm variable is the indicator function of an event $A \subset \Omega$, given by $I_A(\omega) = 1$ if $\omega \in A$ and $I_A(\omega) = 0$ otherwise. If $B$ is any Borel set, then

$$
I_A^{-1}(B) = 
\begin{cases}
A, & \text{if } 1 \in B \text{ and } 0 \notin B,\\
A^c, & \text{if } 1 \notin B \text{ and } 0 \in B,\\
\varnothing, & \text{if } 1 \notin B \text{ and } 0 \notin B,\\
\Omega, & \text{if } 1 \in B \text{ and } 0 \in B.
\end{cases}
$$ 
 
Since all the preimages of $I_A$ are in $\mathcal{F}$, then $I_A$ is a random variable. However, verifying that every possible preimage of $X$ belongs to $\mathcal{F}$  as we did for $I_A$ is impractical. Fortunately, we can check whether $X$ is a random variable by ensuring that this condition holds for a class $\mathcal{C}$ of sets that generate the Borel $\sigma$-algebra on $\mathbb{R}$.

 
**Theorem 2.1** 
 
Let $X: \Omega \to \mathbb{R}$ be a function, and let $\mathcal{C}$ a class of subsets of $\mathbb{R}$ that generates the Borel $\sigma$-algebra. Then, $X$ is a random variable if and only if for every $C \in \mathcal{C}$, we have
 $$X^{-1}(C) = \{ \omega \in \Omega: X(\omega) \in C \} \in \mathcal{F}.$$  

where $\mathcal{F}$ is the $\sigma$-algebra of the sample space $\Omega$.
  
**Proof:**  $\Rightarrow$ Let's define $\mathcal{G}=\{A \in \mathcal{B}(\mathbb{R}): X^{-1}(A) \in \mathcal{F}\}$. It is easy to prove that $\mathcal{G}$ is a $\sigma$-algebra. Since by assumption, $\mathcal{C} \subset \mathcal{G}$, we have that $\mathcal{B}(\mathbb{R}) = \sigma(\mathcal{C}) \subset \mathcal{G}\subset \mathcal{B}(\mathbb{R})$. Therefore, $\mathcal{G}=\mathcal{B}(\mathbb{R})$, and $X$ is a random variable. The other direction is trivial. 

$\blacksquare$ 
 
**Example 2.1** Let $\{X_i\}_{i=1}^n$ a finite collection of random variables respect to $\mathcal{F}$, with $X_i:\Omega\rightarrow\mathbb{R}^n$, then $X=(X_1,X_2,...,X_n)$ is a random variable respect in $\langle \mathcal{F}, \mathcal{B}(\mathbb{R}^n) \rangle$. 
 
Since $\mathcal{B}(\mathbb{R}^n)$ is generated by the class $\mathcal{C}$ of open rectangles  in $\mathbb{R}^n$, then let $C = \prod_{i=1}^n (a_i, b_i)$ be an open rectangle. Then, $\forall C \in \mathcal{C}$, we have
$$X^{-1}(C) = \prod_{i=1}^n X_i^{-1}((a_i, b_i)) = \bigcap_{i=1}^n \{ \omega: X_i(\omega) \in (a_i, b_i)\} \in \mathcal{F}$$ 
 
since $\{ \omega: X_i(\omega) \in (a_i, b_i)\}$ is in $\mathcal{F}$ for all $i=1,2,...,n$ due to the fact that $X_i$ is a random variable. Therefore, $X$ is a random variable by Theorem 1.1.   $\blacksquare$

An useful case is when $\mathcal{C}$ is the class of intervals of the form $(-\infty, x]$. Theorerem 1.1 from the first article states that Borel's $\sigma$-algebra is generated by the class $\mathcal{C} = \{(-\infty, x]: x \in \mathbb{R}\}$. This leads to the following corollary. 

**Corollary 2.2**

Let $X: \Omega \to \mathbb{R}$ be a function. Then, $X$ is a random variable if and only if for every $x \in \mathbb{R}$, we have
$$X^{-1}((-\infty, x]) = \{ \omega \in \Omega: X(\omega) \leq x \} \in \mathcal{F}.$$  

The above corollary is particularly useful to confirm that a function is a random variable, by stating that is sufficient to check if the preimage of intervals of the form $(-\infty, x]$ is in $\mathcal{F}$ as we show in the following example. 
 
 
 **Example 2.2** Let $\{X_n\}_{n\in\mathbb{N}}$ a sequence of random variables in $\langle \mathcal{F}, \mathcal{B}(\mathbb{\bar{R}}) \rangle$. Then, $X = \sup_{n\in\mathbb{N}} X_n$ is a random variable in $\langle \mathcal{F}, \mathcal{B}(\mathbb{\bar{R}}) \rangle$.   
  
For every $x \in \mathbb{R}$, we have
  
$$X^{-1}((-\infty, x]) = \{ \omega: X(\omega) \leq x \} = \bigcap_{n\in\mathbb{N}}\{\omega: X_n(\omega)\leq x\}=\bigcap_{n\in\mathbb{N}} X_n^{-1}((-\infty, x]) \in \mathcal{F}$$ 
 
since $X_n$ are random variables. Therefore, $X$ is a random variable by Corollary 1.2. The same argument can be used for the infimum, taking $\inf_{n\in\mathbb{N}} X_n = - \sup_{n\in\mathbb{N}} (-X_n)$. $\blacksquare$ 
 


  
**Proposition 2.3** Let $X:\Omega\rightarrow\mathbb{R^p}$ be a random variable and $g:\mathbb{R^p}\rightarrow\mathbb{R^q}$ measurable in $\langle \mathcal{B}(\mathbb{R^p}), \mathcal{B}(\mathbb{R^q}) \rangle$. Then, $Y=g(X)$ is a random variable in $\langle \mathcal{F}, \mathcal{B}(\mathbb{R^q}) \rangle$.  
 
**Proof:** Left as an exercise. $\blacksquare$  
 
In order to apply the above proposition safely, we need to know when a real function is measurable. A sufficient condition is given by the following proposition.
 
**Proposition 2.4** Let $g:\mathbb{R^p}\rightarrow\mathbb{R^q}$ be continuous. Then, $g$ is measurable in $\langle \mathcal{B}(\mathbb{R^p}), \mathcal{B}(\mathbb{R^q}) \rangle$.
 
**Proof:** Left as an exercise. Hint: use the fact that the preimage of an open set under a continuous function is open. $\blacksquare$ 
 
 **Example 2.3** If we have $n$ random variables $\{X_k\}_{k=1}^n$ in $\langle \mathcal{F}, \mathcal{B}(\mathbb{R}) \rangle$, then $Y = X_1 + X_2 + ... + X_n$ is a random variable in $\langle \mathcal{F}, \mathcal{B}(\mathbb{R}) \rangle$. 
  
To see this, recall by **Exampple 1.3** that $X=(X_1, X_2, ..., X_n)$ is a random variable in $\langle \mathcal{F}, \mathcal{B}(\mathbb{R}^n) \rangle$. Also, the function $g(x_1, x_2, ..., x_n) = x_1 + x_2 + ... + x_n$ is continuous in $\mathbb{R}^n$ and therefore measurable, then by **Proposition 1.5**, $Y=g(X) = X_1 + X_2 + ... + X_n$ is a random variable in $\langle \mathcal{F}, \mathcal{B}(\mathbb{R}) \rangle$. $\blacksquare$  
 
Another important examples of random variables are presented in the following proposition. The reader should be able to prove the following proposition by himself. 

**Proposition 2.5** Let $X_n:\Omega\rightarrow\mathbb{R}$ be a random variable $\forall n\in\mathbb{N}$, then the following functions are random variables: 

i. $\sup_{n\in\mathbb{N}} X_n$ 

ii. $\inf_{n\in\mathbb{N}} X_n$ 

iii. $\limsup_{n\in\mathbb{N}} X_n$ 

iv. $\liminf_{n\in\mathbb{N}} X_n$ 


**Proof:** Left as an exercise. $\blacksquare$


 ## Distribution of a random variable
 Recall that we use random variables because most of the time we are interested in a quantitative characterization of the outcomes instead of the outcomes themselves. A random variable allows us to abstract ourselves from the original sample space and simply work with the Borel sets. We know that $X^{-1}(B)$ is measurable for every $B \in \mathcal{B}(\mathbb{R})$, so we can define a new probability measure on $\mathcal{B}(\mathbb{R})$ as follows: 
  
$$
P_X(B) = P(X^{-1}(B)) = P(\{ \omega \in \Omega: X(\omega) \in B \}) = P(X\in B)
$$

 It is easy to check that $P_X$ is a probability measure on $\mathcal{B}(\mathbb{R})$.  

 **Definition 2.3 (Distribution of a random variable)** Let $X$ be a random variable on $(\Omega, \mathcal{F}, P)$. The distribution of $X$ is the probability measure $P_X$ on $\mathcal{B}(\mathbb{R})$. 
  
Some examples of distributions are:  

i. $P(X=k) = \displaystyle\frac{1}{k(k+1)}$ for $k=1,2,3,...$. Note that $\sum_{k=1}^{\infty} P(X=k) = 1$ since it is a telescoping series starting at $k=1$. 

ii. Let $B$ be a Borel set, $P(X\in B) = \displaystyle\frac{2}{\pi} \int_{B\cap\mathbb{R}^+} \frac{1}{1+x^2} dx$. Take into account that $\displaystyle\int_{0}^{\infty} \frac{1}{1+x^2} dx = \frac{\pi}{2}$.

 So we can see that any random variable induces a probability measure on $\mathcal{B}(\mathbb{R})$. An important use of this concept through the cumulative distribution function (CDF) of $X$, defined as
 
 $$
 F_X(x) = P_X((-\infty, x]) = P(X\leq x), \quad x \in \mathbb{R}
 $$ 
  
**Example 2.4 (Cumulative effect of $F_X$)** Let $X$ be a random variable on $\Omega = \{1,2,3\}$ with $P(X=1) = 0.1$, $P(X=2) = 0.3$ and $P(X=3) = 0.6$. Then, some values of the CDF of $X$ are given by
 
 - $F_X(a) = 0$ for $a < 1$
 - $F_X(1) = 0.1$
 - $F_X(2) = 0.1 + 0.3 = 0.4$
 - $F_X(3) = 0.1 + 0.3 + 0.6 = 1$
 - $F_X(b) = 1$ for $b > 3$ 
  
  $\blacksquare$  
   
**Proposition 2.6 (Properties of the CDF)** Let $F_X$ be the CDF of a random variable $X$. Then,

i. $F_X$ is non-decreasing.

ii. $F_X$ is right-continuous, i.e., $\lim_{x\to x_0^+} F_X(x) = F_X(x_0)$ for all $x_0 \in \mathbb{R}$.

iii. $\lim_{x\to-\infty} F_X(x) = 0$ and $\lim_{x\to\infty} F_X(x) = 1$.

**Proof:** 
i. Let $x_1 < x_2$. Then, $(-\infty, x_1] \subset (-\infty, x_2]$. Therefore, $F_X(x_1) = P(X\leq x_1) \leq P(X\leq x_2) = F_X(x_2)$.

ii. Let $x_n \downarrow x_0$. Then, $(-\infty, x_n] \downarrow (-\infty, x_0]$. Therefore, by the continuity of $P$, we have $F_X(x_n) \downarrow F_X(x_0)$.

iii. Let $x_n \to -\infty$. Then, $(-\infty, x_n] \to (-\infty, -\infty] = \varnothing$. Therefore, $F_X(x_n) \to 0$. Let $x_n \to \infty$. Then, $(-\infty, x_n] \to (-\infty, \infty) = \mathbb{R}$. Therefore, $F_X(x_n) \to 1$.


$\blacksquare$ 
 
 By continuity of $P$ from below, the reader should be able to prove that $P(X<x) = F_X(x-)$ for all $x \in \mathbb{R}$. Since $\{X\le x\} = \{X<x\} \cup \{X=x\}$, we have $P(X\le x) = P(X<x) + P(X=x)$. Therefore, $P(X=x) = P(X\le x) - P(X<x) = F_X(x) - F_X(x-)$. This way, we see that $F_X$ has jumps at the points $x$ where $P(X=x) > 0$. From real analysis, a monotone function can have at most a countable number of discontinuities. Therefore, the set
 $\{x: P(X=x) > 0\}$ is countable. So, naturally we were led to the following definitions.

**Definition 2.4 (Discrete random variable)** A random variable $X$ is called discrete if its distribution is concentrated on a countable set, i.e., if there exists a countable set $A \subset \mathbb{R}$ such that $P(X \in A) = 1$.
 
 **Definition 2.5 (Continuous random variable)** A random variable $X$ is called continuous if $P(X=x) = 0$ for all $x \in \mathbb{R}$.  
  
Next code shows plots of the CDF of a continuous and a discrete random variable example that we will see later.
  
 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, bernoulli

x_cont = np.linspace(-4, 4, 500)
x_disc = np.array([-1, 0, 1, 2])

# Continuous CDF: Normal distribution
mu, sigma = 0, 1
cdf_cont = norm.cdf(x_cont, mu, sigma)

# Discrete CDF: Bernoulli(p)
p = 0.7
cdf_disc = bernoulli.cdf(x_disc, p)


fig, axs = plt.subplots(1, 2, figsize=(12, 4))

axs[0].plot(x_cont, cdf_cont, color='green')
axs[0].set_title('Continuous CDF')
axs[0].set_xlabel('x')
axs[0].set_ylabel('F(x)')
axs[0].grid(True)

axs[1].step(x_disc, cdf_disc, where='post', color='green')
axs[1].set_title('Discrete CDF')
axs[1].set_xlabel('x')
axs[1].set_ylabel('F(x)')
axs[1].set_xticks(x_disc)
axs[1].grid(True)

plt.tight_layout()
plt.show()
```

## Integration of random variables: Expectation

From school we all have an intuitive understanding of the "average", formally known as the arithmetic mean. Our final score was computed as an average of our scores in the different subjects, that way future reviewer of our academic performance could have an idea about how well(or bad) we did. In the end, the average
$$
\bar{x} = \frac{1}{n} \sum_{i=1}^n x_i
$$
is a way to summarize the dataset into a single, representative number, aiming to capture the central tendency of that data. Now, suppose that there are values repeated in the dataset, let's say that the value $x_i$ occurs $n_i$ times in a dataset of size $m\ge n$. Then, the average can be rewritten as
$$
\bar{x} = \frac{1}{m} \sum_{i=1}^n n_i x_i = \sum_{i=1}^n \frac{n_i}{m} x_i
$$
where $\frac{n_i}{m}$ is the relative frequency of the value $x_i$ in the dataset. This way, the average can be seen as a weighted sum of the values in the dataset, where the weights are the relative frequencies of the values. In the long run, we can replace the relative frequencies by the probabilities of the values $x_i$ which can be interpreted as the values of a discrete random variable $X$. If we set $A_i=\{X=x_i\}$, then $X=\sum_{i=1}^n x_i \mathbb{1}_{A_i}$. Variables of this form are called **simple** as we define next, assuming that we have a probability space $(\Omega, \mathcal{F}, P)$. 

> **Definition 2.4 (Simple random variable)** A random variable $X$ is called simple if it can be written as $X=\sum_{i=1}^n x_i \mathbb{1}_{A_i}$, where $x_i \in \mathbb{R}$ and $A_i \in \mathcal{F}$ for all $i=1,2,...,n$.






### 2. The Expectation as an Integral

#### 2.1  Simple random variables

> **Definition 2.5 (Expectation of a simple random variable)**  
> Let \(X=\sum_{i=1}^{n}x_i\mathbf1_{A_i}\) be a simple r.v. on \((\Omega,\mathcal F,P)\).  
> Its **expected value** (Lebesgue integral) is
> \[
> \boxed{\;
> \mathbb E[X] \;=\;
> \int_{\Omega}X\,\mathrm dP
> := \sum_{i=1}^{n}x_i\,P(A_i)
> \;}
> \]

This mirrors the “weighted average” idea from the introduction: each value \(x_i\) is weighted by the probability of the event where it occurs.

---

#### 2.2  Non-negative random variables

A general non-negative r.v. need not be simple, but it can be **approximated from below** by an increasing sequence of simple ones.

> **Definition 2.6 (Expectation of a non-negative r.v.)**  
> For \(X\ge 0\) measurable, set
> \[
> \boxed{\;
> \mathbb E[X]
> \;:=\;
> \sup\bigl\{
> \mathbb E[S]:
> 0\le S\le X,\;S\ \text{simple}
> \bigr\}.
> \;}
> \]

Equivalently (and often more concrete), choose a sequence \((S_k)_{k\ge1}\) of simple r.v.’s such that  
\(0\le S_1\le S_2\le\cdots\uparrow X\) point-wise and define  
\[
\mathbb E[X]=\lim_{k\to\infty}\mathbb E[S_k]\qquad\text{(monotone convergence).}
\]

---

#### 2.3  Integrable random variables

Any real-valued r.v. \(X\) can be split into its positive and negative parts:
\[
X^+:=\max\{X,0\},\qquad
X^-:=\max\{-X,0\},\qquad
X = X^{+}-X^{-}.
\]

> **Definition 2.7 (Integrable random variable)**  
> \(X\) is **integrable** if \(\mathbb E[X^{+}] + \mathbb E[X^{-}] <\infty\).  
> Then we set
> \[
> \boxed{\;
> \mathbb E[X] := \mathbb E[X^{+}] - \mathbb E[X^{-}].
> \;}
> \]

This matches the classical “finite first moment” requirement.

---

#### 2.4  The discrete random variable as a special case

Suppose \(X\) takes **countably** many values \(\{x_i\}_{i=1}^{\infty}\subset\mathbb R\).
Write \(A_i:=\{X=x_i\}\).  Then

\[
X = \sum_{i=1}^{\infty}x_i\,\mathbf1_{A_i},
\qquad
\mathbf1_{A_i}\in\mathcal F.
\]

Although the sum is infinite, all but one indicator is zero on each \(\omega\), so the series converges point-wise.  
Define the partial-sum simples \(S_n(\omega):=\sum_{i=1}^{n}x_i\mathbf1_{A_i}(\omega)\); clearly \(0\le S_1\le S_2\le\dots\uparrow X^+\) (handle \(X^-\) analogously if needed).

Applying the monotone-convergence-based definition,

\[
\mathbb E[X]
=
\lim_{n\to\infty}\sum_{i=1}^{n}x_iP(A_i)
=
\sum_{i=1}^{\infty}x_i\,P(X=x_i),
\]

provided \(\sum_i|x_i|P(X=x_i)<\infty\).  
Thus the abstract integral collapses to the familiar **counting-measure sum**.

---

### 2.5  Worked examples

| Random variable \(X\) | PMF / PDF | Using the definition above | Result |
|-----------------------|-----------|---------------------------|--------|
| **Bernoulli\(p\)** | \(P(X=1)=p,\;P(X=0)=1-p\) | \(\mathbb E[X]=1\cdot p+0\cdot(1-p)\) | \(p\) |
| **Poisson\(\lambda\)** | \(P(X=k)=e^{-\lambda}\lambda^k/k!\) | \(\sum_{k=0}^{\infty}k\,P(X=k)=\lambda\) | \(\lambda\) |

---

#### 2.6  Intuition checkpoint

1. **Average re-framed:**  the arithmetic mean used counts \(n_i/m\); the Lebesgue integral upgrades those counts to full-blown probabilities on an abstract sample space.  
2. **Why “supremum”?**  For arbitrary \(X\ge0\) we **may not** have an obvious sequence of step functions; taking the supremum over all simple r.v.’s guarantees a well-defined value and leads naturally to Monotone Convergence and Fatou’s Lemma.  
3. **Compatibility:**  whenever a classical Riemann-style formula makes sense (e.g.\ finite discrete support or nice densities), the Lebesgue definition reproduces it exactly.

---

> **Key takeaway:**  *Expectation is nothing more (and nothing less) than a Lebesgue integral with respect to the probability measure.*  
> Starting from simple building blocks and an “average” intuition, we gain a definition powerful enough to handle everything from dice rolls to Itô integrals.












## Discrete random variable distributions
 
 The probability distribution $P_X$ of a discrete random variable has into account that the range of $X$ partitions the sample space $\Omega$ into disjoint events, this is, $\Omega = \bigcup_{x\in A} \{X=x\}$, where $P(X\in A) = 1$, $|A| < \infty$. The analytical expression of $P_X$ often is deduced from combinatorial reasoning when the sample space $\Omega$ is finite, otherwise we need to use more sophisticated tools from analysis.  

### Examples of discrete random variable distributions
**Bernoulli distribution** 
We say that $X$ has a Bernoulli distribution with parameter $p$ if $P(X=1) = p$ and $P(X=0) = 1-p$. The CDF of $X$ is given by

$$
F_X(x) = 
\begin{cases}
0, & \text{if } x < 0,\\
1-p, & \text{if } 0 \leq x < 1,\\
1, & \text{if } x \geq 1,\\
\end{cases}
$$

Note that Bernoulli distribution can also be stated as $P(X=x) = p^x (1-p)^{1-x}$ for $x \in \{0,1\}$. It is trivial that $P(X=1)+P(X=0)=1$. The experiment that leads to this distribution is the so called **Bernoulli trial**, which is a random experiment with two possible outcomes, usually called success and failure. An example of this experiment is the toss of a coin, where success is defined as "head" and failure as "tail". If we make $n$ independent Bernoulli trials, we may be interested in the probability of getting exactly $k$ successes. To compute this probability, we use combinatorics. There are $\binom{n}{k}$ ways to choose $k$ outcomes that are successes out of $n$ trials, and each success has probability $p$ of occurring, while each failure has probability $1-p$ of occurring. Therefore, the probability of getting exactly $k$ successes is given by the expression in the following definition.

 
**Binomial distribution** 

Let $X$ be the number of successes in $n$ independent Bernoulli trials with success probability $p$. Then, $X$ has a binomial distribution with parameters $n$ and $p$, denoted by $X \sim \text{Binomial}(n,p)$. The distribution of $X$ is given by

$$
P(X=k) = \binom{n}{k} p^k (1-p)^{n-k}
$$ 

It is not difficult to show that the sum of the probabilities of all possible values of $X$ is 1, i.e., $\sum_{k=0}^n P(X=k) = 1$. This is a consequence of the binomial theorem applied to the expansion of $(p+(1-p))^n$.
 
 **Example 2.5** 7 coins are tossed independently, and the number of heads is recorded. What is the probability of getting exactly $3$ heads if every coin has probability $p=3/5$ of landing heads? 

Let $X$ be the number of heads in the $7$ tosses. Then, $X$ has a binomial distribution with parameters $n=7$ and $p=3/5$, i.e., $X \sim \text{Binomial}(7,3/5)$. The probability of getting exactly $3$ heads is given by

$$
P(X=3) = \binom{7}{3} (3/5)^3 (2/5)^{7-3} = 0.1166
$$

 $\blacksquare$ 

 The CDF of the binomial distribution is given by
 
 $$
 F_X(x) = \sum_{k \in \mathbb{N}_0:k\leq x} \binom{n}{k} p^k (1-p)^{n-k}
 $$
  
By checking the behavior of the distribution as variable of $k$, we can stablish some interesting facts. Basic computation shows that 
$$
\frac{P(X=k+1)}{P(X=k)} = \frac{n-k}{k+1} \frac{p}{q}
$$

where $q=1-p$. Comparing this ratio with $1$, we see that 
 
i. $P(X=k+1) > P(X=k)$ if and only if $k<np-q$.

ii. $P(X=k+1) < P(X=k)$ if and only if $k > np-q$.
 
iii. $P(X=k+1) = P(X=k)$ if and only if $k = \lfloor np-q \rfloor$. If $np-q$ is an integer, the equality shows that the maximum is attained at two points, namely $k=np-q$ and $k'=np-q+1$.

Therefore, the distribution has its maximum at $k = \lfloor np-q \rfloor$(and $k'=np-q+1$ if $np-q$ is an integer).


We may encounter situations where the number of trial is large, and the computation of the binomial coefficient may be cumbersome.  Fortunately, there are ways to approximate the binomial distribution. To this end, there are two predominant cases regarding the value of $p$:

i. $p$ is fixed independent of $n$. In this case, the distribution is symmetric and the binomial distribution can be approximated by the normal distribution as we'll see later.

ii. $p$ is close to $0$ for big $n$. In this case, the distribution is skewed and the binomial distribution can be approximated by the Poisson distribution as we describe when we introduce the Poisson distribution next. 
 
**Poisson distribution**

In the context of the binomial distribution, when the occurrence of certain event is rare for a huge number of trials, the Poisson distribution is a suitable approximation. 
 
 **Theorem 2.7** Let $X\sim \text{Binomial}(n,p_n)$, and let $\lambda_n = np_n$. If $\lim_{n\to\infty} \lambda_n = \lambda \in (0,\infty)$, then
 
 $$
 \lim_{n\to\infty} P(X=k) = \frac{\lambda^k}{k!} e^{-\lambda}
 $$

 **Proof:** From definition of binomial coefficient and basic results from real analysis, we have: 
  
  1. $\lim_{n\to\infty} \binom{n}{k}/n^k = \frac{1}{k!}$

  2. $\lim_{n\to\infty} \left(1+\frac{x_n}{n}\right)^n = e^{\lim_{n\to\infty} x_n}$.  
  
 Then, from the definition of binomial distribution, we have: 
 
 \begin{align*}
 P(X=k) &= \binom{n}{k} p_n^k (1-p_n)^{n-k} = \binom{n}{k} \left(\frac{\lambda_n}{n}\right)^k \left(1-\frac{\lambda_n}{n}\right)^{n-k}\\
 &= \bigg(\binom{n}{k}/n^k\bigg) \lambda_{n}^k \left(1-\frac{\lambda_n}{n}\right)^{n-k}\\
 &= \bigg(\binom{n}{k}/n^k\bigg) \lambda_{n}^k \frac{\left(1-\frac{\lambda_n}{n} \right)^{n}}{\left(1-\frac{\lambda_n}{n}\right)^k}\\
 \end{align*}
 
 Now, we can apply the limit to both sides of the equation, 
 
 \begin{align*}
 \lim_{n\to\infty} P(X=k) &= \lim_{n\to\infty} \bigg(\binom{n}{k}/n^k\bigg) \lambda_{n}^k \frac{\left(1-\frac{\lambda_n}{n} \right)^{n}}{\left(1-\frac{\lambda_n}{n}\right)^k}\\
 &= \frac{\lambda^k}{k!} e^{-\lambda}
 \end{align*} 
  
$\blacksquare$

**Definition 2.6 (Poisson distribution)** A random variable $X$ has a Poisson distribution with parameter $\lambda > 0$ if its distribution is given by

$$
P(X=k) = \frac{\lambda^k}{k!} e^{-\lambda}
$$


The reader can easily verify that $\sum_{k=0}^\infty P(X=k) = 1$, so the distribution is a valid probability distribution.

Note that in the process of proving the theorem, we took $\lambda = np$, which can be considered as the average(or expected) number of occurrences of the event in the $n$ trials for a binomial distribution. 
 
**Example 2.6** A bank has 900 borrowers, and the probability that a borrower will default is $p=0.01$. What is the probability that exactly 7 borrowers will default?

Let $X$ be the number of borrowers that will default. Then, $X$ has a binomial distribution with parameters $n=900$ and $p=0.01$. The probability that exactly 7 borrowers will default is given by

$$
P(X=7) = \binom{900}{7} (0.01)^7 (0.99)^{900-7} \approx 0.1173
$$

but this is a very difficult computation, so we can use the Poisson distribution to approximate the binomial distribution. The average number of defaults is $\lambda = np = 900 \times 0.01 = 9$. Therefore, the probability that exactly 7 borrowers will default is given by

$$
P(X=7) = \frac{9^7}{7!} e^{-9} \approx 0.1171
$$

We see that the Poisson approximation is very good, and the computation is much easier.

$\blacksquare$

**Example 2.7** A particular stock has an average of 60 trades per hour. If a trader steps away from the trading desk for 30 seconds, what is the probability that there will be no trades during that interval?

Let $X$ be the number of trades in the interval of 30 seconds. Since $1h=3600s = 120\times 30s$, then there are 120 intervals of 30 seconds in an hour. The average number of trades in an hour is 60, so the average number of trades in 30 seconds is $\lambda = 60/120 = 0.5$. Therefore, $X\sim \text{Poisson}(0.5)$. The probability that there will be no trades during that interval is given by

$$
P(X=0) = \frac{0.5^0}{0!} e^{-0.5} = e^{-0.5} \approx 0.6065
$$

$\blacksquare$ 
 
**Geometric distribution**

Consider a sequence of independent Bernoulli trials with success probability $p$. We want to know the probability that the first success occurs at the $k$-th trial. The answer is given by the following definition.

**Definition 2.7 (Geometric distribution)** A random variable $X$ has a geometric distribution with parameter $p$, denoted by $X\sim \text{Geometric}(p)$, if its distribution is given by

$$
P(X=k) = (1-p)^{k-1} p
$$ 
 
The above equation follows from the fact that the probability of the first success at the $k$-th trial is the probability of $k-1$ failures followed by a success. Using the definition of geometric series, we can check that $\sum_{k=1}^\infty P(X=k) = 1$, so the distribution is a valid probability distribution.  
 
**Example 2.8** In a simple trading algorithm, the trade can make money with estimated probability $p=0.4$. If the trader keeps running the algorithm until the first loss appears, what is the probability that the first loss appears at the 5th trade?

Let $X$ be the number of trades until the first loss appears. Then, $X\sim \text{Geometric}(0.4)$, so the required probability is given by

$$
P(X=5) = (1-0.4)^{5-1} \times 0.4 = 0.6^4 \times 0.4 = 0.05184
$$

$\blacksquare$
